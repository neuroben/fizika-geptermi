### Pár gondolat a NumPy-ről

NumPy: Numerical Python modul. Szokásos hívása: `import numpy as np`.
  * https://numpy.org/
  * https://en.wikipedia.org/wiki/NumPy

Nagyon sokat tudó modul, mely lehetőséget ad Pythonban (relatíve) gyors numerikus számításokra. Megoldás arra, hogy a Python listák (rugalmasságuk miatt) nem hatékonyak, ha sok, azonos típusú értéket tárolunk bennük és ezekkel végzünk műveletet.

Számumnkra fontos NumPy alap típusok:
   * `np.float32`, `np.float64`: 32 illetve 64 bites lebegőpontos szám
   * `np.int32`, `np.int64`: 32 illetve 64 bites egész
   * `np.bool_`: logikai érték (True vagy False)

A NumPy alap adat-struktúrája a tömb (array), amik csupa azonos típusú elemből állnak. Főbb sajátosságok:
   * a listákhoz hasonlóan indexelhetők
   * támogatják az úgynevezett "vektorizált" végrehajtást, azaz amikor operátor vagy függvény automatikusan végrehajtódik minden elemre; pl. ha `arr1` és `arr2` két, azonos méretű NumPy tömb, akkor `arr1-arr2` egy ugyanekkora tömb, elemenként kiszámolva a különbséget
   * sok művelet egy paranccsal elérhető, pl. `arr.mean()` a tömbelemek átlaga, `arr.std()` a szórás (standard deviation)
   * sok algoritmus, pl. regresszió, interpoláció, egyenlet-megoldás, be van programozva
   
A használat többnyire elég kézenfekvő. Érdemes megtanulni, mert sokkal tömörebb és gyorsabb a kód, mintha Python list-ekkel és ciklusokkal dolgoznánk. (De egy C++-ban írt és lefordított program sebességét nem éri el.)

Sok más projekt épít erre, pl. SciKit-Learn, SciPy (Scientific Python), OpenCV (Open Computer Vision), Pandas, ...

# 1D NumPy tömbök (array-ek)

## NumPy array-ek létrehozása

```python
# adott lista elemeiből: 
# np.array(lista, dtype=típus)
aa=np.array([0.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0], dtype=np.float64)
print(aa)
```

```python
# csupa 0 érték, adott méret és típus
# np.zeros(méret, dtype=típus)
bb=np.zeros(7, dtype=np.int32)
print(bb)
```

```python
# csupa 0 érték, másolt méret és típus
# np.zeros_like(minta_tömb)
cc=np.zeros_like(aa) # aa-ból veszi a méretet és a típust
print(cc)
```

```python
# Egyenletes lépésközű feltöltés:

# np.arange(start, end, delta, dtype=típus)
dd=np.arange(0.0, 1.0, 0.1, dtype=np.float64) 
print("np.arange kimenet:\t", dd)
# 'end' érték már nem szerepel a listában!

# np.linspace(start, end, number, dtype=típus)
ee=np.linspace(0.0, 1.0, 11, dtype=np.float64) 
print("np.linspace kimenet:\t", ee)
# 'end' érték is szerepel a listában!
```

## Lekérdezések

```python
# A tömb fő paraméterei:
print("A tömb elemei:", aa)
print("A tömb mérete:", aa.shape)
print("A tömbelemek típusa:", aa.dtype)
```

## Matematikai műveletek

```python
# Bármilyen egyszerű függvény:
aa_new=aa*3
print("aa*3 =\t", aa_new)

dd_new=10*dd**3-8
print("10*dd**3-8=\t",dd_new)
```

```python
# Matematikai függvények, csak pár példa
# fokból radiánba:
aa_rad=np.radians(aa)  
print("aa_rad=\t", aa_rad)

# szögfüggvények:
dd_sin=np.sin(dd)
dd_arctan=np.arctan(dd)
print("dd_sin=\t", dd_sin)
print("dd_arctan=\t", dd_arctan)

# kombinálva:
aa_rad_sin=np.sin(np.radians(aa))
print("aa_rad_sin=\t", aa_rad_sin)


# exponenciális függvény:
dd_exp=np.exp(dd)
print("dd_exp=\t", dd_exp)
```

```python
# Saját függvény vektorizálása (hogy egy lépésben alkalmazható legyen NumPy array-re)
# np.vectorize(myfun)

def myfun(x):
    if x<0.5:
        return 2*x
    else:
        return 9*x**2
    
myfun_vec=np.vectorize(myfun)
# 'myfun_vec': vektorizált változat

# egy elemre ugyanolyan, mint az eredeti:
print(f"myfun(0.8)={myfun(0.8):.6f}; myfun_vec(0.8)={myfun_vec(0.8):.6f}")

# tömbre már csak a vektorizált alkalmazható:
print('myfun_vec(ee)',myfun_vec(ee))

# tömbre a nem vektorizált hibát ad:
print('myfun(ee)',myfun(ee))
```

## Statisztikai függvények: pár példa

```python
# gyakorló tömb:
ff=np.array([0.01, 1.08, 2.02, 0.52, -0.33, -1.12, 0.0])
print("ff elemei:\t", ff)
```

```python
# Összegző függvények
ff_sum=ff.sum()
ff_mean=ff.mean()
ff_std=ff.std()
print(f"ff elemeinek átlaga: {ff_mean:.6f}, összege: {ff_sum:.6f}, szórása: {ff_std:.6f}")
```

```python
# Szélső érték keresése:
ff_min=ff.min()
ff_max=ff.max()
print(f"ff elemeinek minimuma: {ff_min:.6f}, maximuma: {ff_max:.6f}")
```

```python
# Szélső érték helyének keresése:

ff_indmin=ff.argmin()
ff_indmax=ff.argmax()

print(f"ff legkisebb elemének indexe: {ff_indmin}; legnagyobb elemének indexe: {ff_indmax}")
```

## Kombinált keresések

```python
# teszt tömbök:
t=np.arange(0.0, 3.0, 0.2, dtype=np.float64)
x=12+3*t-2*t**2

print("t=\t",t)
print("x=\t",x)
```

```python
# Mennyi 't' azon eleme, ami 'x' maximumához tartozik?

ind_xmax=x.argmax()  
t_xmax=t[ind_xmax]
print(f"'x' maximumának indexe: {ind_xmax}; az ennek megfelelő elem 't'ből: {t_xmax:.6f}")
```

```python
# trükkösebb: szomszédok különbsége

t_nemelso=t[1:]   # t[1], t[2], ..., azaz lemarad az 0-s sorszámú
print("nem első elemek:", t_nemelso)
print("nem első elemek száma:", t_nemelso.shape)
```

```python
t_nemutolso=t[:-1]   # t[0], t[1], ..., t[utolsó előtti] azaz lemarad az utolsó, azaz a -1-es sorszámú
print("nem utolsó elemek:", t_nemutolso)
print("nem utolsó elemek száma:", t_nemutolso.shape)
```

```python
delta_t=t_nemelso-t_nemutolso # t szomészédos elemeinek különbsége
print(delta_t)
```

```python
# hasonlóan a delta_x-ek, de már tömören:
delta_x=x[1:]-x[:-1]

print(delta_x)
```

```python
# mivel ez gyakran kell, ezért van rá egy rutin:
delta_x=np.diff(x)

print(delta_x)
```

```python
# átlagsebesség szakaszról szakaszra:
v_atl=delta_x/delta_t
print(v_atl)

# Vigyázat! Ez eggyel kevesebb elemű, mint x!
```

```python
# ezek legjobban az intervalumok közepén levő sebességeket közelítik
t_kozep=(t[1:]+t[:-1])/2
print(t_kozep)
```

```python
fig=plt.figure(figsize=(10,4)) 
ax1=fig.add_subplot(111) 
ax1.set_xlabel("t [s]")     
ax1.set_ylabel("v(t) [m/s]")     
ax1.grid()              

ax1.plot(t_kozep, v_atl, color="orange", label="v(t)")   # t függvényében v_atl

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

```python
# Figyelem! delta_t, delta_x, v_atl egyel rövidebb, mint t és x!
```

```python
# út = kis elmodulások nagyságának összege
teljes_ut=np.abs(delta_x).sum()   # tömb.sum(): tömbelemek összege
print(f"Teljes út: {teljes_ut:.3f} m.")
```

```python
# Ha az összegzés részeredményei is érdekesek: tömb.cumsum()
ut=np.abs(delta_x).cumsum()
print(ut)
```

```python
# Ebből kimarad a 0. elem, de addig az út nyilván 0 volt. Ezért ez a korrekt:
ut=np.append([0], np.abs(delta_x).cumsum())  # append: összefűzés
print(ut)
```

```python
fig=plt.figure(figsize=(10,4)) 
ax1=fig.add_subplot(111) 
ax1.set_xlabel("t [s]")     
ax1.set_ylabel("ut(t) [m]")      
ax1.grid()               

ax1.plot(t, ut, color="red", label="x(t)")   # t függvényében ut

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

## Az új eljárások gyűjteménye

Tömb létrehozás:
   * `aa=np.array(lista, dtype=típus)`: adott lista elemeit a típus-ra alakítja és ebből hoz létre tömböt
   * `aa=np.zeros(N, dtype=típus)`: `N` méretű, csupa 0-ból álló tömb
   * `bb=np.zeros_like(aa)`: ugyanakkora és ugyanolyan típusú tömb létrehozása, mint `aa`
   * `dd=np.append(aa, bb)`: két tömb egymás után fűzése és tárolása
   * `cc=np.arange(start, end, delta, típus)`: `start` és `end` közt `delta` egyenletes lépésekkel létrejövő tömb
   * `cc2=np.linspace(start, end, N, típus)`: `start` és `end` közt egyenletes lépésekkel létrejövő `N` elemű tömb

Matemetikai függvények:
   * Műveleti jelek: `+`, `-`, `*`, `/`, `**`, ...: NumPy tömbökre elemenként vonatkoznak
   * `np.radians()`: fokról radiánra vált
   * `np.sin()`, `np.cos()`, `np.abs()`, ...: szinusz, koszinusz, abszolút érték, ...
   * `vfun=np.vectorize(fun)`: előállítja a `fun` vektorizált változatát

Összegzés:
   * `aa.sum()`: a tömb összes elemének összege
   * `aa.diff()`: szomszédok különbségéből álló tömb
   * `aa.cumsum()`: egy tömb fokozatos felösszegzése (kumulatív szumma)

Keresés:
   * `aa.max()`, `aa.min()`: a tömb maximális/minimális elemének nagysága
   * `aa.argmax()`, `aa.argmin()`: a tömb maximális/minimális elemének indexe (hányadik helyen van)
