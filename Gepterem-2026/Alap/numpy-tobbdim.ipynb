# Többdimenziós tömbök kezelése a NumPy-ben

Ez a notebook bevezetés a NumPy többdimenziós tömb kezelő rutinjaiba. A példák mind kétdimenziós tömbökre vonatkoznak, de hasonlóan akárhány dimenzióra is működnének. 

Itt csak azokat a funkciókat ismertetjük, melyeket felhasználunk a későbbiekben a tárgyból.

```python
import numpy as np    
import matplotlib.pyplot as plt
%matplotlib inline    
```

## 2D NumPy tömbök létrehozása

```python
# létrehozás listák listájából
aa_list= [ [1, 2, 3, 4] ,
           [5, 6, 7, 8] 
         ]

aa=np.array(aa_list, dtype=np.float64)

print("A tömb elemei :\n", aa)
print("A tömb méretei:",aa.shape)
```

```python
# 0-val feltöltött tömb létrehozása
zz=np.zeros((2,3), np.int32)
print(zz)
```

```python
# 0-val feltöltött, másik tömbhöz hasonló létrehozása:
zzaa=np.zeros_like(aa)
print(zzaa)
```

```python
# komplett másolat létrehozása:
cc=aa.copy()
print(cc)
```

```python
# sorfolytonos megadásból 2d tömb a reshape paranccsal:
dd=np.reshape([10,20,30,40,50,60], (2,3))
print(dd)
```

```python
# 2d tömbből sorfolytonossá alakítás:
ss=dd.flatten()
print(ss)
```

## Méret lekérdezés

```python
print("aa méretei:", aa.shape)
print("dd méretei:", dd.shape)
```

```python
print("aa sorainak száma:", aa.shape[0])
print("aa oszlopainak száma:", aa.shape[1])
```

```python
print("aa elemeinek száma:", aa.size)
```

```python
# A len csak a sorok számát (0. dimenzió) adja meg:
print("aa sorainak száma:", len(aa))
print("dd sorainak száma:", len(dd))
```

## Műveletek 2d NumPy tömbökkel

```python
# Az egyszerű vektorizálás könnyen működik:
bb=aa**2+3
print(bb)
```

```python
# Lehet csak bizonyos sorokra vagy oszlopokra hivatkozni
# Első, azaz 0-s sorszámú sor, minden oszlop
print("Első sor:\n",bb[0,:])
# utolsó oszlop:
print("Utolsó oszlop:\n",bb[:,-1])
```

```python
# értékadással is működik:
# minden második oszlop nullázása
bb[:,::2]=0
print(bb)
```

```python
hh1=np.reshape([8.0,-1,4,8,-4,5], (2,3))
hh2=np.reshape([-8.0,+3,2,7,0,5], (2,3))

print("Két teszt-tömb:")
print(hh1)
print(hh2)
```

```python
# elemenkénti összegek azonos méretű tömbö közt:
hh_sum=hh1+hh2
print(hh_sum)
```

```python
# elemenkénti maximumok:
print(np.maximum(hh1, hh2))
```

```python
# Az egyik elem lehet egy skalár is:

print("hh1 pi-szerese:")
print(np.pi*hh1)

print("hh1 nemnegatív elemei:")
print(np.maximum(hh1, 0))
```

## Az axis-ek koncepciója

```python
# NumPy többdimenziós tömbök: tengelyek mentén rendezett adatok.
# 2D tömb: 0-s axis=sor, 1-es axis=oszlop
# számolási rutinok vonatkozhatnak csak bizonyos axis-ekre
```

```python
# gyakorló tömb:
aa_list= [ 
           [1, 2, 3, 4] ,
           [5, 6, 7, 8] 
         ]

aa=np.array(aa_list, dtype=np.float64)
```

```python
# Minden elem összege: egy szám
print(aa.sum())
```

```python
# összegzés a 0-s axis mentén: annyi szám, ahány oszlop, sorok "összeejtése"
print(aa.sum(axis=0))
```

```python
# összegzés a 1-s axis mentén: annyi szám, ahány sor, oszlopok "összeejtése"
print(aa.sum(axis=1))
```

```python
# sorok átlaga
print(aa.mean(axis=1))
```

```python
# oszlopok négyzetösszegéből vont gyök:
print((aa**2).sum(axis=0)**0.5)
# Ezt használni fogjuk vektorok abszolút értékének számolásakor
```

## Boolean értékek felhasználása

Az 1D esetben tanultakhoz hasonlóan.

```python
# aa elemei oszthatók-e 3-mal?
map_3n= (aa%3==0)
print(map_3n)
# kimenet: boolean "térkép" arról, hol igaz a feltétel
```

```python
# Csak a 3-mal sztható elemek print-elése:
print(aa[map_3n])
```

```python
# Ilyen boolean tömb felhasználható értékadásra is:
# A 3-mal osztható elemeket elosztjuk 3-mal
aa[map_3n] /= 3 
print(aa)
```

```python
# Ahol True van, azok tömbje megkapható az argwhere paranccsal:
np.argwhere(map_3n)
```

```python
# hol nagyobb hh1, mint hh2?
hh1_dominates=hh1>hh2
print(hh1_dominates)
```

```python
print("hh1:\n", hh1)
print("hh2:\n", hh2)

print("hh1 nagyobb elemet tartalmaz, mint hh2 az alábbi pozíciókban:")
for arg in np.argwhere(hh1_dominates):
    print(arg)
```

## Gyakorló feladatok

Mivel ebben nincs elvi újdonság a korábban tanultakhoz képest, megadom mindegyik megoldását. A fizika példákra való alkalmazásokban lesz idő a saját feladatmegoldásra. Mostani cél: mindegyiket megérteni.

```python
ff_list=[[0, 1], [1, 4], [2, 5], [3, 9]]
ff=np.array(ff_list, np.float64)
print(ff)
```

**1. feladat:**

Írja ki, hogy ff hány sorból áll!

```python
print("Sorok száma:",ff.shape[0])
```

**2. feladat**

Adja meg ff sorainak összegét!

```python
ff.sum(axis=1)
```

**3. feladat**

Ha ff nem más, mint 4 db 2 dimenziós vektor, akkor mi ezen vektorok hossza?

```python
print(((ff**2).sum(axis=1))**0.5)
```

**4. feladat**

Ha ff nem más, mint 4 db 2 dimenziós vektor, akkor melyik ezek közül a legnagyobb hosszúságú?

```python
ff_abs=((ff**2).sum(axis=1))**0.5
print("A legnagyobb absz. értékű vektor indexe:",ff_abs.argmax())
```

**5. feladat**

ff mely soraiban fordul elő, hogy az 1-es komponens nagyobb 4-nél?

```python
print("A következő sorokban nagyobb az 1-es komponens 4-nél:")
print(np.argwhere(ff[:,1]>4).flatten())
```

**6. feladat**

ff mely soraiban fordul elő, hogy az 1-es komponens nagyobb a 0-s kétszeresénél?

```python
print("A következő sorokban nagyobb az 1-es komponens a 0-s kétszeresénél:")
print(np.argwhere(  ff[:,1] > ff[:,0]*2  ).flatten())
```

**7. feladat**

Nullázza ff azon elemeit, melyek nagyobbak ff átlagának felénél!

```python
average=ff.mean()

ff[ ff>average/2 ] = 0

print(ff)
```

# Új NumPy eljárások

   * `aa.size`: tömb elemeinek száma

   * `np.reshape(tömb_vagy_lista, shape)`: a megadott tömb vagy lista elemeből a `shape`-nek megfelelő átrendezést hoz létre
