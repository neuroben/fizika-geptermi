## 1. Fizikai Kérdéstípusok


### 1.1 Sebességnagyság ábrázolása idő és útvonalhossz szerint

**Vizsgaszöveg:** Ábrázolja a vonat sebességének nagyságát az idő és az útvonalhossz függvényében is! / Ábrázolja az autó sebességének nagyságát az idő és az útvonalhossz függvényében is!

**Fizikai jelentés:** A sebesség nagysága azt mutatja, milyen gyorsan halad a test, függetlenül attól, merre mozog. Az idő szerinti ábra azt mutatja, mikor gyors vagy lassú a mozgás, az útvonalhossz szerinti ábra pedig azt, hogy a pálya mely részein történt ez. A kód azért működik, mert a `v_abs` a sebességvektor nagysága, a `pathlength` pedig a pályán addig megtett út.

**Minimális kódrecept:** A következő kódcella kiszámítja a sebességnagyságot és két grafikonon ábrázolja.

**Gyakori hiba:** A sebesség egyik komponensét rajzolják ki a sebesség nagysága helyett, pedig kanyarodó mozgásnál a komponens akár előjelet is válthat.

**Feladatok:** Gy/1a, Gy/3a.


```python
t = np.arange(0.0, 220.0, 0.2)
sebesseg = 18.0 + 6.0*np.sin(2*np.pi*t/90.0) + 2.0*np.sin(2*np.pi*t/25.0)
x = integ(t, sebesseg)
y = 120.0*np.sin(x/900.0)

jarmu = num_kinem(2)
jarmu.t = t
jarmu.r = np.column_stack((x, y))
jarmu.full_kinem_calc()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(jarmu.t, jarmu.v_abs)
ax1.set_xlabel('idő (s)')
ax1.set_ylabel('sebességnagyság (m/s)')
ax1.grid()

ax2.plot(jarmu.pathlength, jarmu.v_abs)
ax2.set_xlabel('útvonalhossz (m)')
ax2.set_ylabel('sebességnagyság (m/s)')
ax2.grid()

plt.show()
```

### 1.2 Mennyi ideig volt a test álló helyzetben?

**Vizsgaszöveg:** Állónak tekintjük azokat a szakaszokat, amikor a sebesség nagysága 1 m/s-nál kisebb. Hány másodpercig állt a vonat összesen?

**Fizikai jelentés:** A vonat nem matematikailag pontosan nulla sebességnél számít állónak, hanem egy kis küszöb alatt, mert a GPS-adatok és a numerikus deriválás zajosak lehetnek. Azért a feltételhez tartozó időlépéseket adjuk össze, mert minden ilyen kis intervallumban a vonat a feladat definíciója szerint áll. Így a teljes állásidő a sok kis álló időszakasz összege.

**Minimális kódrecept:** A következő kódcella `v_abs < 1` maszkkal kiválasztja az álló időlépéseket, majd összeadja a hozzájuk tartozó `dt` értékeket.

**Gyakori hiba:** A True értékek darabszámát adják meg másodpercként, pedig az időlépések hosszát kell összegezni.

**Feladatok:** Gy/1b.


```python
t = np.arange(0.0, 420.0, 0.5)
sebesseg = 16.0 + 3.0*np.sin(2*np.pi*t/180.0)
sebesseg[(100.0 <= t) & (t < 150.0)] = 0.2
sebesseg[(285.0 <= t) & (t < 330.0)] = 0.3
x = integ(t, sebesseg)
y = np.zeros_like(x)

vonat = num_kinem(2)
vonat.t = t
vonat.r = np.column_stack((x, y))
vonat.full_kinem_calc()

allo_hatar = 1.0
dt = vonat.t[1:] - vonat.t[:-1]
allt = vonat.v_abs[:-1] < allo_hatar
allasido = dt[allt].sum()

print(f'A vonat összesen {allasido:.1f} s ideig állt.')
```

### 1.3 Mennyi ideig teljesült egy nagysebességű feltétel, és hány szakasz volt?

**Vizsgaszöveg:** Mennyi ideig haladt a vonat a maximális sebesség 80%-ánál nagyobb sebességgel és hány ilyen nagy-sebességű szakasz volt?

**Fizikai jelentés:** A maximális sebességhez viszonyított küszöb azt jelöli ki, mikor tekintjük a mozgást nagysebességűnek. Az időtartam azért a küszöb feletti időlépések összege, mert csak ezekben az intervallumokban teljesül a fizikai feltétel. A szakaszok száma nem a pontok száma, hanem az összefüggő küszöb feletti tartományok száma.

**Minimális kódrecept:** A következő kódcella megkeresi a `v_abs > 0.8*vmax` pontokat, összeadja az időlépéseket, és megszámolja az új True blokkok kezdetét.

**Gyakori hiba:** A küszöb feletti adatpontok számát keverik össze a nagysebességű szakaszok számával.

**Feladatok:** Gy/1c.


```python
t = np.arange(0.0, 600.0, 0.5)
sebesseg = 13.0 + 7.0*np.sin(2*np.pi*t/170.0)**2 + 1.5*np.sin(2*np.pi*t/55.0)
x = integ(t, sebesseg)
y = 40.0*np.sin(x/1000.0)

vonat = num_kinem(2)
vonat.t = t
vonat.r = np.column_stack((x, y))
vonat.full_kinem_calc()

kuszob = 0.8 * vonat.v_abs.max()
dt = vonat.t[1:] - vonat.t[:-1]
gyors = vonat.v_abs[:-1] > kuszob
gyors_ido = dt[gyors].sum()
uj_szakasz = gyors & np.r_[True, ~gyors[:-1]]
szakaszok_szama = uj_szakasz.sum()

print(f'Küszöb: {kuszob:.2f} m/s')
print(f'Küszöb feletti összidő: {gyors_ido:.1f} s')
print(f'Nagysebességű szakaszok száma: {szakaszok_szama}')

plt.figure(figsize=(9, 4))
plt.plot(vonat.t, vonat.v_abs, color='black')
plt.scatter(vonat.t[:-1][gyors], vonat.v_abs[:-1][gyors], color='red', s=5)
plt.axhline(kuszob, color='red', linestyle='--')
plt.xlabel('idő (s)')
plt.ylabel('sebességnagyság (m/s)')
plt.grid()
plt.show()
```

### 1.4 Erős fékezési és erős kanyarodási szakaszok jelölése

**Vizsgaszöveg:** Készítsen ábrát, melyen a vonat pályája szürke, és rajta pirossal az erős fékezési (0.8 m/s^2-nél erősebb), kékkel az erős kanyarodási (0.7 m/s^2-nél erősebb) szakaszok vannak bejelölve!

**Fizikai jelentés:** Fékezéskor a gyorsulás sebességirányú komponense negatív, mert a gyorsulás a mozgás irányával ellentétesen hat. Kanyarodáskor a sebesség iránya változik, ezt a centripetális gyorsulás nagysága méri. A kód ezért `a_t_abs < -küszöb` feltétellel jelöli a fékezést és `a_cp_abs > küszöb` feltétellel a kanyarodást.

**Minimális kódrecept:** A következő kódcella két fizikai feltételmaszkot készít, majd a pályán külön színnel jelöli őket.

**Gyakori hiba:** Fékezéshez `a_abs`-t használnak, pedig az nem mondja meg, hogy a gyorsulás a sebesség irányába vagy azzal ellentétesen mutat.

**Feladatok:** Gy/1d, Vizsga-minta-01/F2b.


```python
t = np.arange(0.0, 180.0, 0.1)
sebesseg = 18.0 + 10.0*np.sin(2*np.pi*t/70.0)
szogsebesseg = 0.006 + 0.055*np.exp(-((t - 85.0)/13.0)**2)
iranyszog = integ(t, szogsebesseg)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

vonat = num_kinem(2)
vonat.t = t
vonat.r = np.column_stack((x, y))
vonat.full_kinem_calc()

fek_hatar = 0.8  # Erős fékezés határa (pálya menti lassulás mértéke) [m/s^2]
kan_hatar = 0.7  # Erős kanyarodás határa (centripetális gyorsulás mértéke) [m/s^2]

# Logikai maszkok (True/False tömbök) a feltételeket teljesítő pontok kiválasztásához
# A fékezésnél azért van mínusz jel, mert az 'a_t_abs' lassuláskor negatív értéket vesz fel!
eros_fekezes = vonat.a_t_abs < -fek_hatar  
eros_kanyar = vonat.a_cp_abs > kan_hatar

plt.figure(figsize=(10, 4))
plt.plot(vonat.r_x, vonat.r_y, color='lightgray', label='pálya')
plt.scatter(vonat.r_x[eros_fekezes], vonat.r_y[eros_fekezes], color='red', s=6, label='erős fékezés')
plt.scatter(vonat.r_x[eros_kanyar], vonat.r_y[eros_kanyar], color='blue', s=6, label='erős kanyarodás')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Vizsga-minta-01/F2b kiegészítés: egy térképen egyszerre jelöljük
# a nagyon lassú és az erősen fékező GPS-szakaszokat.
t = np.linspace(0.0, 80.0, 801)
v = 11.0 + 2.0*np.sin(2*np.pi*t/32.0)

mask = (t >= 18) & (t < 25)
v[mask] = np.linspace(11.0, 0.4, mask.sum())
mask = (t >= 25) & (t < 32)
v[mask] = 0.4
mask = (t >= 32) & (t < 45)
v[mask] = np.linspace(0.4, 15.0, mask.sum())
mask = (t >= 55) & (t < 61)
v[mask] = np.linspace(15.0, 2.5, mask.sum())
v = np.maximum(v, 0.2)

irany = 0.20*np.sin(2*np.pi*t/55.0)
vx = v*np.cos(irany)
vy = v*np.sin(irany)

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack([integ(t, vx), integ(t, vy)])
auto.full_kinem_calc()

lassu = auto.v_abs < 1.0
eros_fekezes = auto.a_t_abs < -0.7

plt.figure(figsize=(8, 4))
plt.plot(auto.r_x, auto.r_y, color='0.75', label='pálya')
plt.scatter(auto.r_x[lassu], auto.r_y[lassu], s=12, color='green', label='v < 1 m/s')
plt.scatter(auto.r_x[eros_fekezes], auto.r_y[eros_fekezes], s=12, color='red', label='a_t < -0,7 m/s²')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

### 1.5 Analitikusan megadott pálya ábrázolása

**Vizsgaszöveg:** Ábrázolja a motor pályáját az első 100 másodpercben!

**Fizikai jelentés:** A pálya a test helyének sorozata a síkban, nem az idő függvényében rajzolt külön koordináta. Ha a feladat megadja az `x(t)` és `y(t)` függvényt, akkor minden időponthoz tartozik egy helyvektor. A kód azért működik, mert `set_r_fun` ezekből a helyvektorokból tölti fel a mozgást, a grafikon pedig az `x-y` pontokat köti össze.

**Minimális kódrecept:** A következő kódcella létrehozza a motor helyfüggvényét, kiszámítja a kinematikai mennyiségeket, majd kirajzolja a pályát.

**Gyakori hiba:** Külön `x(t)` és `y(t)` időgrafikont rajzolnak, pedig a kérdés a síkbeli pályagörbét kéri.

**Feladatok:** Gy/2a, Gy/13a, Gy/14a, 25_05_21/Gy8a, F2-2026/a, Vizsga-minta-01/F3a.


```python
def motor_r(t):
    phi = (2*np.pi/12.0) * t
    R = 10.0 + 1.0*np.sin(2*np.pi*t/2.5)
    x = R*np.cos(phi)
    y = R*np.sin(phi)
    return np.array([x, y], np.float64)

motor = num_kinem(2)
motor.set_time_range(0.0, 100.0, 0.01)
motor.set_r_fun(motor_r)
motor.full_kinem_calc()

plt.figure(figsize=(6, 6))
plt.plot(motor.r_x, motor.r_y, color='black')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.grid()
plt.show()
```

```python
# Gy/4a kiegészítés: itt nem helyfüggvény, hanem sebességfüggvény adott.
# Fizikailag a pályát a sebesség idő szerinti integrálásával kapjuk meg.
def jarmu_v(t):
    if t < 3.0:
        return np.array([5.0, 0.0], np.float64)
    tau = t - 3.0
    phi = (2*np.pi/12.0) * tau
    v = 5.0 + 2.3*np.sin(2*np.pi*tau/7.0)
    return np.array([v*np.cos(phi), v*np.sin(phi)], np.float64)

jarmu = num_kinem(2)
jarmu.set_time_range(0.0, 100.0, 0.01)
jarmu.set_v_fun(jarmu_v)
jarmu.calc_v_to_r(r0=np.array([0.0, 0.0]))
jarmu.calc_v_to_a()
jarmu.calc_delta_r()
jarmu.calc_delta_r_abs()
jarmu.calc_pathlength()
jarmu.calc_at_acp_Rinv()

plt.figure(figsize=(7, 7))
plt.plot(jarmu.r_x, jarmu.r_y, color='black')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Pálya sebességfüggvényből integrálva')
plt.grid()
plt.show()
```

```python
# Gy/6a kiegészítés: dinamikai modellből rajzolunk pályát a becsapódásig.
# Új elem: a számítás stop feltétellel áll meg, amikor a labda eléri a talajt.
m_labda = 0.30
v0 = 16.0
r0 = np.array([0.0, 1.8])

def tomeg(t):
    return m_labda

def gravitacio_valtozo(t, hely, sebesseg, m):
    g = 1.625 if t < 5.0 else 9.81
    return np.array([0.0, -m*g], np.float64)

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

szog = np.radians(45.0)
labda = num_dinam(2)
labda.set_time_param(0.0, 40.0, 0.02)
labda.set_mass_fun(tomeg)
labda.set_F_fun(gravitacio_valtozo)
labda.set_r0_v0(r0, np.array([v0*np.cos(szog), v0*np.sin(szog)]))
labda.set_stop_cond(talajt_ert)
labda.full_dinam_calc()

plt.figure(figsize=(8, 4))
plt.plot(labda.r_x, labda.r_y, color='black')
plt.axhline(0.0, color='gray', linestyle='--')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Ferde hajítás pályája becsapódásig')
plt.grid()
plt.show()
```

```python
# Gy/7a kiegészítés: önvezető jármű pályája sebességnagyság-sapkával és változó irányszöggel.
# Új elem: a sebességnagyság gyorsul, majd 30 m/s-nál telítődik.
def onvezeto_v(t):
    if t < 2.0:
        return np.array([10.0, 0.0], np.float64)
    v = min(10.0 + 0.5*(t - 2.0), 30.0)
    phi = np.radians(20.0) * np.sin(2*np.pi*(t - 2.0)/12.0)
    return np.array([v*np.cos(phi), v*np.sin(phi)], np.float64)

jarmu = num_kinem(2)
jarmu.set_time_range(0.0, 100.0, 0.02)
jarmu.set_v_fun(onvezeto_v)
jarmu.calc_v_to_r(r0=np.array([0.0, 0.0]))
jarmu.calc_v_to_a()
jarmu.calc_delta_r()
jarmu.calc_delta_r_abs()
jarmu.calc_pathlength()
jarmu.calc_at_acp_Rinv()

plt.figure(figsize=(9, 4))
plt.plot(jarmu.r_x, jarmu.r_y, color='black')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Önvezető jármű pályája az első 100 s alatt')
plt.grid()
plt.show()
```

```python
# Gy/8a kiegészítés: vízrakéta pályája dinamikai rakétamodellel a becsapódásig.
# Új elem: időfüggő tömeg, tolóerő és közegellenállás szerepel az erőtörvényben.
def raketa_modell():
    szaraz_tomeg = 0.10
    hajtoanyag_tomeg = 0.30
    kiaramlasi_seb = 35.0
    tomegaram = 0.20
    C = 0.20
    sugar = 0.03
    A = np.pi*sugar**2
    rho = 1.30
    g = 9.81

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
            if t <= hajtoanyag_tomeg/tomegaram:
                F += tomegaram*kiaramlasi_seb*sebesseg/v
        return F

    return ero, tomeg

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

ero, tomeg = raketa_modell()
szog = np.radians(85.0)
raketa = num_dinam(2)
raketa.set_time_param(0.0, 40.0, 0.02)
raketa.set_mass_fun(tomeg)
raketa.set_F_fun(ero)
raketa.set_r0_v0(np.array([0.0, 1.0]), np.array([np.cos(szog), np.sin(szog)]))
raketa.set_stop_cond(talajt_ert)
raketa.full_dinam_calc()

plt.figure(figsize=(6, 6))
plt.plot(raketa.r_x, raketa.r_y, color='black')
plt.axhline(0.0, color='gray', linestyle='--')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Vízrakéta pályája becsapódásig')
plt.grid()
plt.show()
```

```python
# Gy/9a kiegészítés: pattogó labda pályája az első 20 másodpercben.
# Új elem: a talajjal való rugalmas kölcsönhatás miatt a pálya több ívből áll.
def pattogo_labda_modell():
    m = 0.59
    kerulet = 0.76
    sugar = kerulet/(2*np.pi)
    A = np.pi*sugar**2
    C = 0.47
    rho = 1.30
    D = 4000.0
    F_f = 3.0
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        if hely[1] < 0.0:
            F[1] += D*(-hely[1])
            if abs(sebesseg[0]) > 1e-10:
                F[0] -= F_f*np.sign(sebesseg[0])
        return F

    return ero, tomeg

ero, tomeg = pattogo_labda_modell()
szog = np.radians(40.0)
labda = num_dinam(2)
labda.set_time_param(0.0, 20.0, 0.003)
labda.set_mass_fun(tomeg)
labda.set_F_fun(ero)
labda.set_r0_v0(np.array([0.0, 1.9]), np.array([10.0*np.cos(szog), 10.0*np.sin(szog)]))
labda.set_stop_cond(None)
labda.full_dinam_calc()

labda.clip_below_y()  # ← ez levágja a talaj alatti pontokat

plt.figure(figsize=(10, 4))
plt.plot(labda.r_x, labda.r_y, color='black')
plt.axhline(0.0, color='gray', linestyle='--')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Pattogó labda pályája 20 s alatt')
plt.grid()
plt.show()
```

### 1.6 Legnagyobb és legkisebb sebesség meghatározása

**Vizsgaszöveg:** Mennyi a motor legnagyobb és legkisebb sebessége?

**Fizikai jelentés:** A sebesség nagysága a pillanatnyi haladási tempó, ezért a legnagyobb és legkisebb sebesség a `v_abs` szélsőértéke. A kód azért működik, mert a pályából számolt sebességvektor minden időpontra ismert, és a vektor nagyságának maximuma/minimuma közvetlenül a keresett fizikai mennyiség. Ilyenkor nem az `x` vagy `y` irányú komponens szélsőértékét keressük.

**Minimális kódrecept:** A következő kódcella ugyanazt a motorpályát használja, majd `v_abs.max()` és `v_abs.min()` alapján válaszol.

**Gyakori hiba:** A sebességkomponensek minimumát és maximumát keresik, pedig azok irányfüggők és nem a sebesség nagyságát adják.

**Feladatok:** Gy/2b.


```python
def motor_r(t):
    phi = (2*np.pi/12.0) * t
    R = 10.0 + 1.0*np.sin(2*np.pi*t/2.5)
    return np.array([R*np.cos(phi), R*np.sin(phi)], np.float64)

motor = num_kinem(2)
motor.set_time_range(0.0, 100.0, 0.01)
motor.set_r_fun(motor_r)
motor.full_kinem_calc()

i_max = motor.v_abs.argmax()
i_min = motor.v_abs.argmin()

print(f'Legnagyobb sebesség: {motor.v_abs[i_max]:.2f} m/s, idő: {motor.t[i_max]:.2f} s')
print(f'Legkisebb sebesség: {motor.v_abs[i_min]:.2f} m/s, idő: {motor.t[i_min]:.2f} s')
```

### 1.7 Pályaszakaszok jelölése relatív sebességküszöb alapján

**Vizsgaszöveg:** Készítsen ábrát, melyen a motor pályája szürke, és rajta pirossal be vannak jelölve azok a szakaszok, ahol a sebesség nagyobb, mint a maximális sebesség 95%-a, zölddel pedig azok, ahol a sebesség kisebb, mint a minimális sebesség 105%-a!

**Fizikai jelentés:** A kérdés nem abszolút sebességhatárt ad, hanem a mozgás saját maximumához és minimumához viszonyít. Ezért először a teljes mozgásból kell meghatározni a szélsőértékeket, majd ezekből képezni a küszöböket. A pályán történő jelölés azért válaszol a kérdésre, mert megmutatja, térben hol fordult elő a nagyon gyors vagy nagyon lassú haladás.

**Minimális kódrecept:** A következő kódcella kiszámítja a relatív küszöböket, majd két maszkkal színezi a pályát.

**Gyakori hiba:** Fix sebességhatárt választanak, pedig a feladat a saját maximum és minimum százalékát kéri.

**Feladatok:** Gy/2c.


```python
def motor_r(t):
    phi = (2*np.pi/12.0) * t
    R = 10.0 + 1.0*np.sin(2*np.pi*t/2.5)
    return np.array([R*np.cos(phi), R*np.sin(phi)], np.float64)

motor = num_kinem(2)
motor.set_time_range(0.0, 100.0, 0.01)
motor.set_r_fun(motor_r)
motor.full_kinem_calc()

gyors = motor.v_abs > 0.95*motor.v_abs.max()
lassu = motor.v_abs < 1.05*motor.v_abs.min()

plt.figure(figsize=(6, 6))
plt.plot(motor.r_x, motor.r_y, color='lightgray', label='pálya')
plt.scatter(motor.r_x[gyors], motor.r_y[gyors], color='red', s=4, label='gyors')
plt.scatter(motor.r_x[lassu], motor.r_y[lassu], color='green', s=4, label='lassú')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

### 1.8 Közelítőleg egyenes szakaszok jelölése

**Vizsgaszöveg:** Ábrázolja azokat a szakaszokat, ahol a motor közelítőleg egyenes vonalban haladt!

**Fizikai jelentés:** Egyenes haladásnál a sebesség iránya alig változik, ezért a kanyarodáshoz tartozó centripetális gyorsulás kicsi. A `a_cp_abs` éppen az irányváltoztatáshoz szükséges gyorsulás nagysága, ezért kis értékei közel egyenes pályarészeket jelölnek. A kódban egy kis küszöb választja el a majdnem egyenes részeket a kanyarodóktól.

**Minimális kódrecept:** A következő kódcella `a_cp_abs < küszöb` maszkkal jelöli a közel egyenes pályarészeket.

**Gyakori hiba:** A sebesség nagyságát nézik, pedig egy test nagy sebességgel is haladhat közel egyenesen, és lassan is kanyarodhat.

**Feladatok:** Gy/2d.


```python
def motor_r(t):
    phi = (2*np.pi/12.0) * t
    R = 10.0 + 1.0*np.sin(2*np.pi*t/2.5)
    return np.array([R*np.cos(phi), R*np.sin(phi)], np.float64)

motor = num_kinem(2)
motor.set_time_range(0.0, 100.0, 0.01)
motor.set_r_fun(motor_r)
motor.full_kinem_calc()

egyenes = motor.a_cp_abs < 0.5

plt.figure(figsize=(6, 6))
plt.plot(motor.r_x, motor.r_y, color='lightgray', label='teljes pálya')
plt.scatter(motor.r_x[egyenes], motor.r_y[egyenes], color='blue', s=4, label='közel egyenes')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

```python
# Gy/7c kiegészítés: egy ábrán jelöljük az erős kanyarodást és a közel egyenes részeket.
# Új elem: két külön fizikai maszk kerül ugyanarra a pályára.
def onvezeto_v(t):
    if t < 2.0:
        return np.array([10.0, 0.0], np.float64)
    v = min(10.0 + 0.5*(t - 2.0), 30.0)
    phi = np.radians(20.0) * np.sin(2*np.pi*(t - 2.0)/12.0)
    return np.array([v*np.cos(phi), v*np.sin(phi)], np.float64)

jarmu = num_kinem(2)
jarmu.set_time_range(0.0, 100.0, 0.02)
jarmu.set_v_fun(onvezeto_v)
jarmu.calc_v_to_r(r0=np.array([0.0, 0.0]))
jarmu.calc_v_to_a()
jarmu.calc_delta_r()
jarmu.calc_delta_r_abs()
jarmu.calc_pathlength()
jarmu.calc_at_acp_Rinv()

eros_kanyar = jarmu.a_cp_abs > 3.0
kozel_egyenes = jarmu.a_cp_abs < 0.5

plt.figure(figsize=(9, 4))
plt.plot(jarmu.r_x, jarmu.r_y, color='lightgray', label='pálya')
plt.scatter(jarmu.r_x[eros_kanyar], jarmu.r_y[eros_kanyar], color='red', s=5, label='erős kanyarodás')
plt.scatter(jarmu.r_x[kozel_egyenes], jarmu.r_y[kozel_egyenes], color='blue', s=5, label='közel egyenes')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

### 1.9 Menetidő és út százaléka sebességküszöb felett

**Vizsgaszöveg:** A menetidő és az út hány százalékát töltötte az autó 65 km/h-nál nagyobb sebességgel?

**Fizikai jelentés:** Ugyanaz a sebességfeltétel két különböző arányt adhat: időarányt és útarányt. Az időarány azt mutatja, a menetidő mekkora részében volt gyors az autó, az útarány pedig azt, a megtett út mekkora részét tette meg gyorsan. A kód azért működik, mert a `dt` az idődarabokat, a `pathlength` különbségei pedig az útdarabokat adják ugyanazon feltételmaszk alatt.

**Minimális kódrecept:** A következő kódcella ugyanazzal a `v_abs > 65/3.6` maszkkal számít idő- és útarányt.

**Gyakori hiba:** Az időszázalékot és az útszázalékot azonosnak gondolják, pedig nagyobb sebességnél ugyanannyi idő alatt nagyobb út telik el.

**Feladatok:** Gy/3b.


```python
t = np.arange(0.0, 900.0, 0.5)
sebesseg_kmh = 58.0 + 22.0*np.sin(2*np.pi*t/310.0) + 8.0*np.sin(2*np.pi*t/95.0)
sebesseg = np.maximum(sebesseg_kmh/3.6, 2.0)
iranyszog = 0.25*np.sin(2*np.pi*t/500.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack((x, y))
auto.full_kinem_calc()

v_kuszob = 65.0/3.6
dt = auto.t[1:] - auto.t[:-1]
ds = auto.pathlength[1:] - auto.pathlength[:-1]
gyors = auto.v_abs[:-1] > v_kuszob

ido_szazalek = 100.0 * dt[gyors].sum() / (auto.t[-1] - auto.t[0])
ut_szazalek = 100.0 * ds[gyors].sum() / auto.pathlength[-1]

print(f'65 km/h feletti időarány: {ido_szazalek:.1f}%')
print(f'65 km/h feletti útarány: {ut_szazalek:.1f}%')
```

```python
# Gy/5b kiegészítés: itt nem sebességküszöb, hanem közel egyenletes haladás a feltétel.
# Fizikailag akkor közel egyenletes a sebességnagyság, ha a tangenciális gyorsulás nagysága kicsi.
t = np.arange(0.0, 500.0, 0.2)
sebesseg = 12.0 + 2.0*np.sin(2*np.pi*t/180.0) + 0.4*np.sin(2*np.pi*t/18.0)
iranyszog = 0.35*np.sin(2*np.pi*t/260.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack((x, y))
auto.full_kinem_calc()

egyenletes_hatar = 0.1
dt = auto.t[1:] - auto.t[:-1]
ds = auto.pathlength[1:] - auto.pathlength[:-1]
egyenletes = auto.a_t_magnitude[:-1] < egyenletes_hatar

ido_szazalek = 100.0 * dt[egyenletes].sum() / (auto.t[-1] - auto.t[0])
ut_szazalek = 100.0 * ds[egyenletes].sum() / auto.pathlength[-1]

print(f'Közel egyenletes haladás időaránya: {ido_szazalek:.1f}%')
print(f'Közel egyenletes haladás útaránya: {ut_szazalek:.1f}%')
```

### 1.10 Erős gyorsítási és erős fékezési szakaszok jelölése

**Vizsgaszöveg:** Készítsen ábrát, melyen az autó pályája szürke, és rajta pirossal az erős gyorsítási (0.5 m/s^2-nél erősebb), kékkel az erős fékezési (0.5 m/s^2-nél erősebb) szakaszok vannak bejelölve!

**Fizikai jelentés:** A gyorsítás és fékezés a sebesség nagyságának változásáról szól, ezért a pálya menti, azaz tangenciális gyorsulás előjele a döntő. Pozitív `a_t_abs` esetén a sebesség nő, negatív `a_t_abs` esetén csökken. A kód ezért külön maszkot készít az erős gyorsításra és az erős fékezésre, majd ezeket a pályán jelöli.

**Minimális kódrecept:** A következő kódcella `a_t_abs > 0.5` és `a_t_abs < -0.5` feltételekkel jelöli az autó pályáján a gyorsítási és fékezési részeket.

**Gyakori hiba:** Az `a_abs` nagy értékét gyorsításnak veszik, pedig nagy gyorsulás kanyarodásból is származhat úgy, hogy a sebesség nagysága alig változik.

**Feladatok:** Gy/3c.


```python
t = np.arange(0.0, 240.0, 0.1)
sebesseg = 19.0 + 7.0*np.sin(2*np.pi*t/80.0)
iranyszog = 0.35*np.sin(2*np.pi*t/180.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack((x, y))
auto.full_kinem_calc()

kuszob = 0.5
eros_gyorsitas = auto.a_t_abs > kuszob
eros_fekezes = auto.a_t_abs < -kuszob

plt.figure(figsize=(10, 4))
plt.plot(auto.r_x, auto.r_y, color='lightgray', label='pálya')
plt.scatter(auto.r_x[eros_gyorsitas], auto.r_y[eros_gyorsitas], color='red', s=6, label='erős gyorsítás')
plt.scatter(auto.r_x[eros_fekezes], auto.r_y[eros_fekezes], color='blue', s=6, label='erős fékezés')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

### 1.11 Adott úthányad elérésének ideje

**Vizsgaszöveg:** Határozza meg, mennyi időbe telt, amíg az autó megtette a teljes útvonal 25, 50 és 75%-át!

**Fizikai jelentés:** A megtett út időben monoton nő, ezért a teljes út 25, 50 és 75%-a jól meghatározott mérföldkő a mozgásban. A `pathlength` minden időponthoz megadja, addig mennyi út gyűlt össze. A kód azért működik, mert megkeresi, mikor lépi át a `pathlength` a megadott részútértékeket.

**Minimális kódrecept:** A következő kódcella kiszámítja a teljes út 25, 50 és 75%-át, majd `arg_eq` segítségével megkeresi az első elérési időket.

**Gyakori hiba:** Az `x` koordináta 25, 50 és 75%-át keresik, pedig a feladat az útvonal mentén megtett teljes útról szól.

**Feladatok:** Gy/3d.


```python
t = np.arange(0.0, 1200.0, 0.5)
sebesseg = 17.0 + 5.0*np.sin(2*np.pi*t/260.0) + 2.5*np.sin(2*np.pi*t/90.0)
iranyszog = 0.45*np.sin(2*np.pi*t/650.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack((x, y))
auto.full_kinem_calc()

teljes_ut = auto.pathlength[-1]
for hanyad in [0.25, 0.50, 0.75]:
    cel_ut = hanyad * teljes_ut
    indexek = arg_eq(auto.pathlength, cel_ut)
    i = indexek[0]
    print(f'{100*hanyad:.0f}% megtéve: {auto.t[i]:.1f} s, hely: ({auto.r_x[i]:.1f} m, {auto.r_y[i]:.1f} m)')
```

### 1.12 Gyorsulás nagyságának ábrázolása idő szerint

**Vizsgaszöveg:** Ábrázolja az autó gyorsulásának nagyságát az idő függvényében!

**Fizikai jelentés:** A gyorsulás nagysága azt mutatja, milyen erősen változik a sebességvektor, függetlenül attól, hogy a változás gyorsulásból, lassulásból vagy kanyarodásból ered. A kód azért `a_abs` értéket rajzol, mert ez a teljes gyorsulásvektor hossza, nem csak egy komponens. Így a grafikon a mozgás dinamikai „erősségét” mutatja idő szerint.

**Minimális kódrecept:** A következő kódcella sebességfüggvényből pályát és gyorsulást számol, majd `a_abs(t)` grafikont készít.

**Gyakori hiba:** Csak az `a_y` vagy csak az `a_t_abs` értéket rajzolják, pedig a kérdés a teljes gyorsulás nagyságát kéri.

**Feladatok:** Gy/4b.

```python
def jarmu_v(t):
    if t < 3.0:
        return np.array([5.0, 0.0], np.float64)
    tau = t - 3.0
    phi = (2*np.pi/12.0) * tau
    v = 5.0 + 2.3*np.sin(2*np.pi*tau/7.0)
    return np.array([v*np.cos(phi), v*np.sin(phi)], np.float64)

jarmu = num_kinem(2)
jarmu.set_time_range(0.0, 100.0, 0.01)
jarmu.set_v_fun(jarmu_v)
jarmu.calc_v_to_r(r0=np.array([0.0, 0.0]))
jarmu.calc_v_to_a()
jarmu.calc_delta_r()
jarmu.calc_delta_r_abs()
jarmu.calc_pathlength()
jarmu.calc_at_acp_Rinv()

plt.figure(figsize=(9, 4))
plt.plot(jarmu.t, jarmu.a_abs)
plt.xlabel('idő (s)')
plt.ylabel('gyorsulás nagysága (m/s²)')
plt.grid()
plt.show()
```

### 1.13 Minimális téglalap alakú tesztpálya mérete

Fizikailag ez a pálya legkisebb tengelyekkel párhuzamos befoglaló téglalapja. A kód azért `max-min` különbségeket használ, mert ezek adják meg, mekkora tartományt jár be a jármű az egyes koordinátairányokban.

**Minimális kódrecept:** A következő kódcella kiszámítja az `x` és `y` koordináták teljes kiterjedését.

### 1.15 Legközelebbi és legtávolabbi pont a kiindulástól adott idő után

**Vizsgaszöveg:** Határozza meg azokat az időpontokat t = 3 s után, melyekre az autó a legközelebb illetve a legtávolabb került a kiindulás helyétől és jelölje be ezeket a test pályáján!

**Fizikai jelentés:** A kiindulási helytől mért távolság nem ugyanaz, mint a megtett út: ez az origótól vett pillanatnyi távolság. A `dist` minden időpontra ezt a távolságot adja, ezért a `t > 3 s` maszk után ennek minimuma és maximuma jelöli a keresett pontokat. A kód ugyanazzal az indexszel olvassa ki az időt és a helyet, ezért fizikailag ugyanarra az eseményre mutat.

**Minimális kódrecept:** A következő kódcella `dist` alapján megkeresi és bejelöli a t = 3 s utáni legközelebbi és legtávolabbi pontot.

**Gyakori hiba:** A `pathlength` értékét használják, pedig az mindig nő, így nem alkalmas a kiindulástól való közelség keresésére.

**Feladatok:** Gy/4e.

```python
t = np.arange(0.0, 100.0, 0.02)
sebesseg = 5.0 + 2.0*np.sin(2*np.pi*t/7.0)
iranyszog = 2*np.pi*t/12.0
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

jarmu = num_kinem(2)
jarmu.t = t
jarmu.r = np.column_stack((x, y))
jarmu.full_kinem_calc()

utana = jarmu.t > 3.0
helyi_indexek = np.where(utana)[0]
i_kozel = helyi_indexek[jarmu.dist[utana].argmin()]
i_tavol = helyi_indexek[jarmu.dist[utana].argmax()]

print(f'Legközelebb: t = {jarmu.t[i_kozel]:.2f} s, távolság = {jarmu.dist[i_kozel]:.2f} m')
print(f'Legtávolabb: t = {jarmu.t[i_tavol]:.2f} s, távolság = {jarmu.dist[i_tavol]:.2f} m')

plt.figure(figsize=(7, 7))
plt.plot(jarmu.r_x, jarmu.r_y, color='lightgray')
plt.scatter(jarmu.r_x[i_kozel], jarmu.r_y[i_kozel], color='blue', marker='*', s=120, label='legközelebb')
plt.scatter(jarmu.r_x[i_tavol], jarmu.r_y[i_tavol], color='purple', marker='*', s=120, label='legtávolabb')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

### 1.17 Hipotetikus menetidő alsó sebességkorláttal

**Vizsgaszöveg:** Mennyi volt a menetidő és mennyi lett volna, ha az autónak sosem kellett volna 10 km/h-nál jobban lelassítania?

**Fizikai jelentés:** A menetidő az útdarabok és a sebesség kapcsolatából adódik: ugyanazt az utat nagyobb sebességgel rövidebb idő alatt tesszük meg. A hipotetikus kérdésben a pálya útdarabjai változatlanok, csak a túl kicsi sebességeket cseréljük legalább 10 km/h-ra. A kód ezért `dt = ds/v` alapján újraszámolja az egyes szakaszok idejét.

**Minimális kódrecept:** A következő kódcella összehasonlítja az eredeti menetidőt és azt az időt, amely akkor adódna, ha a sebesség nem menne 10 km/h alá.

**Gyakori hiba:** Csak a régi menetidőt írják ki, pedig a „mennyi lett volna” újraszámított, feltételezett mozgást kér.

**Feladatok:** Gy/5d, Vizsga-minta-01/F2c.

```python
t = np.arange(0.0, 700.0, 0.5)
sebesseg = 9.0 + 5.0*np.sin(2*np.pi*t/210.0)
sebesseg[(250.0 < t) & (t < 330.0)] = 1.4
sebesseg = np.maximum(sebesseg, 0.8)
iranyszog = 0.25*np.sin(2*np.pi*t/400.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack((x, y))
auto.full_kinem_calc()

sebesseg_hatar = 10.0/3.6
dt = auto.t[1:] - auto.t[:-1]
ds = auto.pathlength[1:] - auto.pathlength[:-1]
v_eredeti = auto.v_abs[:-1]
v_korrigalt = np.maximum(v_eredeti, sebesseg_hatar)
dt_korrigalt = ds / v_korrigalt

print(f'Eredeti menetidő: {dt.sum():.1f} s')
print(f'Hipotetikus menetidő: {dt_korrigalt.sum():.1f} s')
```

### 1.18 Dobástávolság és maximális magasság

**Vizsgaszöveg:** Mi ilyenkor a dobás távolsága és az elért legnagyobb magasság?

**Fizikai jelentés:** A dobástávolság a becsapódási pont vízszintes távolsága a dobás helyétől, a maximális magasság pedig a pálya legnagyobb `y` koordinátája. A pályahossz nem azonos a dobástávolsággal, mert a labda íves pályán mozog. A kód ezért a vízszintes végkoordinátát (`r_x[-1]`) és a `max_height` értéket olvassa ki.

**Minimális kódrecept:** A következő kódcella 45 fokos hajítást szimulál változó gravitációval, majd kiírja a vízszintes távolságot és a legnagyobb magasságot.

**Gyakori hiba:** `pathlength[-1]`-et használnak dobástávolságként, pedig az a görbe pálya hossza, nem a vízszintes lőtávolság.

**Feladatok:** Gy/6b, Gy/14e.

```python
m_labda = 0.30
v0 = 16.0
r0 = np.array([0.0, 1.8])

def tomeg(t):
    return m_labda

def gravitacio_valtozo(t, hely, sebesseg, m):
    g = 1.625 if t < 5.0 else 9.81
    return np.array([0.0, -m*g], np.float64)

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

szog = np.radians(45.0)
labda = num_dinam(2)
labda.set_time_param(0.0, 40.0, 0.02)
labda.set_mass_fun(tomeg)
labda.set_F_fun(gravitacio_valtozo)
labda.set_r0_v0(r0, np.array([v0*np.cos(szog), v0*np.sin(szog)]))
labda.set_stop_cond(talajt_ert)
labda.full_dinam_calc()

print(f'Dobástávolság: {labda.r_x[-1]:.2f} m')
print(f'Legnagyobb magasság: {labda.max_height:.2f} m')
```

```python
def raketa_szimulacio(hajtoanyag_tomeg=0.35, szaraz_tomeg=0.15,
                     v_kiaramlas=45.0, tomegaram=0.2, sugar=0.04,
                     alaktenyezo=0.2, szog_fok=85.0, kezdo_v=1.0,
                     kezdo_y=1.0, dt=0.02, t_max=25.0):
    rho = 1.2
    g = 9.81
    A = np.pi * sugar**2
    egesi_ido = hajtoanyag_tomeg / tomegaram

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5 * alaktenyezo * A * rho * v * sebesseg
            if t <= egesi_ido:
                F += tomegaram * v_kiaramlas * sebesseg / v
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, t_max, dt)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(
        np.array([0.0, kezdo_y]),
        np.array([kezdo_v*np.cos(szog), kezdo_v*np.sin(szog)])
    )
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

# Gy/14e kiegészítés: előbb megkeressük, melyik szögnél maximális a repülés közbeni
# sebességnagyság, majd ugyanennél a szögnél olvassuk ki a magasságot és a távolságot.
szogek = np.linspace(20.0, 88.0, 35)
max_sebessegek = []

for szog in szogek:
    raketa = raketa_szimulacio(szog_fok=szog, dt=0.03, t_max=18.0)
    repulesben = raketa.r_y >= 0.0
    max_sebessegek.append(raketa.v_abs[repulesben].max())

max_sebessegek = np.array(max_sebessegek)
optimalis_szog = szogek[max_sebessegek.argmax()]
optimalis_raketa = raketa_szimulacio(szog_fok=optimalis_szog, dt=0.02, t_max=18.0)

print(f'Optimális szög a sebességmaximumhoz: {optimalis_szog:.1f} fok')
print(f'Ehhez tartozó maximális magasság: {optimalis_raketa.r_y.max():.2f} m')
print(f'Ehhez tartozó lőtávolság: {optimalis_raketa.r_x[-1]:.2f} m')
```

### 1.19 Kilövési szög optimalizálása maximális távolságra

**Vizsgaszöveg:** Milyen irányt válasszon a 45 fokos helyett, hogy a labda a legmesszebb repüljön?

**Fizikai jelentés:** A kilövési szög meghatározza, mennyi kezdősebesség jut vízszintes haladásra és mennyi függőleges emelkedésre. A legnagyobb távolság ott adódik, ahol ez a két hatás optimálisan oszlik meg. A kód azért szögeken söpör végig, mert minden szögre ugyanazt a fizikai mozgást számolja ki, majd a legnagyobb `r_x[-1]` értékhez tartozó szöget választja.

**Minimális kódrecept:** A következő kódcella több eldobási szöget kipróbál, és megkeresi, melyik adja a legnagyobb dobástávolságot.

**Gyakori hiba:** Automatikusan 45 fokot választanak, pedig változó gravitáció vagy nem szokványos kezdőmagasság esetén az optimális szög eltérhet ettől.

**Feladatok:** Gy/6c.

```python
m_labda = 0.30
v0 = 16.0
r0 = np.array([0.0, 1.8])

def tomeg(t):
    return m_labda

def gravitacio_valtozo(t, hely, sebesseg, m):
    g = 1.625 if t < 5.0 else 9.81
    return np.array([0.0, -m*g], np.float64)

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

def hajitas(szog_fok):
    szog = np.radians(szog_fok)
    labda = num_dinam(2)
    labda.set_time_param(0.0, 40.0, 0.03)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(gravitacio_valtozo)
    labda.set_r0_v0(r0, np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

szogek = np.linspace(5.0, 85.0, 81)
tavolsagok = np.array([hajitas(szog).r_x[-1] for szog in szogek])
i_max = tavolsagok.argmax()

print(f'Optimális eldobási szög: {szogek[i_max]:.1f} fok')
print(f'Legnagyobb dobástávolság: {tavolsagok[i_max]:.2f} m')

plt.figure(figsize=(8, 4))
plt.plot(szogek, tavolsagok)
plt.scatter(szogek[i_max], tavolsagok[i_max], color='red', zorder=3)
plt.xlabel('eldobási szög (fok)')
plt.ylabel('dobástávolság (m)')
plt.grid()
plt.show()
```

### 1.20 Minimális és maximális eldobási szög akadály feletti átrepüléshez

**Vizsgaszöveg:** Mennyi a minimális és maximális eldobási szög, ha a labdának át kell repülnie egy, a dobótól 30 m-re levő, 40 m magas fal felett?

**Fizikai jelentés:** A fal feletti átrepülés nem a teljes pálya maximumától függ, hanem attól, hogy a labda a fal vízszintes helyén milyen magasan van. Egy szög akkor megfelelő, ha az `x = 30 m` helyen a pálya `y` értéke nagyobb 40 m-nél. A kód ezért sok szöget kipróbál, mindegyik pályán ellenőrzi az akadályfeltételt, majd a megfelelő szögek minimumát és maximumát adja meg.

**Minimális kódrecept:** A következő kódcella szögsöpréssel megkeresi azokat az eldobási szögeket, amelyekkel a labda átrepül a fal felett.

**Gyakori hiba:** Csak azt nézik, hogy a maximális magasság nagyobb-e 40 m-nél, pedig a labdának éppen a fal helyén kell elég magasan lennie.

**Feladatok:** Gy/6d.

```python
m_labda = 0.30
v0 = 16.0
r0 = np.array([0.0, 1.8])
fal_x = 30.0
fal_y = 40.0

def tomeg(t):
    return m_labda

def gravitacio_valtozo(t, hely, sebesseg, m):
    g = 1.625 if t < 5.0 else 9.81
    return np.array([0.0, -m*g], np.float64)

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

def hajitas(szog_fok):
    szog = np.radians(szog_fok)
    labda = num_dinam(2)
    labda.set_time_param(0.0, 40.0, 0.03)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(gravitacio_valtozo)
    labda.set_r0_v0(r0, np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

jo_szogek = []
for szog in np.linspace(1.0, 89.0, 177):
    labda = hajitas(szog)
    atrepul, y_ertekek = labda.flies_over_obstacle(fal_x, fal_y)
    if atrepul:
        jo_szogek.append(szog)

jo_szogek = np.array(jo_szogek)
print(f'Minimális megfelelő szög: {jo_szogek.min():.1f} fok')
print(f'Maximális megfelelő szög: {jo_szogek.max():.1f} fok')
```

```python
# Gy/10e kiegészítés: több kezdőirány is jó lehet, ha adott x-nél adott y magasságot kell eltalálni.
# Új elem: nem maximumot keresünk, hanem azokat a szögeket, ahol a pálya átmegy az ablak magasságán.
def teniszlabda(szog_fok):
    m = 0.058
    A = np.pi*(0.067/2)**2
    C = 0.55
    rho = 1.225
    g = 9.81
    v0 = 15.0

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    labda = num_dinam(2)
    labda.set_time_param(0.0, 8.0, 0.006)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.8]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

fal_x = 8.0
ablak_y = 4.5
talalatok = []

for szog in np.linspace(5.0, 85.0, 321):
    labda = teniszlabda(szog)
    if labda.passes_near_y_at_x(fal_x, ablak_y, 0.05):
        talalatok.append(szog)

print('Ablakot közel eltaláló kezdőszögek:')
for szog in talalatok:
    print(f'{szog:.2f} fok')
```

### 1.21 Leállás adott megtett út után

**Vizsgaszöveg:** Ha egy automatika leállítja a mozgást akkor, amikor az autó 2 km utat megtett az indulástól kezdve, akkor hol és mikor áll le?

**Fizikai jelentés:** A kérdés nem adott x-koordinátát, hanem a pálya mentén összegyűlt megtett utat ad meg. A `pathlength` monoton növekszik, ezért a 2000 m első elérése egyértelmű leállási eseményt jelöl. A kód azért működik, mert a célúthoz tartozó indexszel ugyanabból az időpillanatból olvassa ki az időt és a helyet.

**Minimális kódrecept:** A következő kódcella megkeresi a `pathlength = 2000 m` átlépését, majd kiírja a hozzá tartozó időt és koordinátákat.

**Gyakori hiba:** Az `r_x` koordinátát keresik 2000 m-nél, pedig a feladat a megtett útról, nem a vízszintes elmozdulásról szól.

**Feladatok:** Gy/7d.

```python
def onvezeto_v(t):
    if t < 2.0:
        return np.array([10.0, 0.0], np.float64)
    v = min(10.0 + 0.5*(t - 2.0), 30.0)
    phi = np.radians(20.0) * np.sin(2*np.pi*(t - 2.0)/12.0)
    return np.array([v*np.cos(phi), v*np.sin(phi)], np.float64)

jarmu = num_kinem(2)
jarmu.set_time_range(0.0, 100.0, 0.02)
jarmu.set_v_fun(onvezeto_v)
jarmu.calc_v_to_r(r0=np.array([0.0, 0.0]))
jarmu.calc_v_to_a()
jarmu.calc_delta_r()
jarmu.calc_delta_r_abs()
jarmu.calc_pathlength()
jarmu.calc_at_acp_Rinv()

cel_ut = 2000.0
i = arg_eq(jarmu.pathlength, cel_ut)[0]

print(f'Leállási idő: {jarmu.t[i]:.2f} s')
print(f'Leállási hely: x = {jarmu.r_x[i]:.2f} m, y = {jarmu.r_y[i]:.2f} m')
print(f'Megtett út ekkor: {jarmu.pathlength[i]:.2f} m')
```

### 1.22 Teljes út és elmozdulás nagysága

**Vizsgaszöveg:** Mekkora a mozgás során a rakéta teljes útja és elmozdulásának nagysága?

**Fizikai jelentés:** A teljes út a pályagörbe mentén összeadott ívhossz, az elmozdulás pedig csak a kezdőpontból a végpontba mutató egyenes vektor hossza. A két mennyiség görbe pályán nem azonos. A kód azért `pathlength[-1]`-et használ teljes útnak, az elmozduláshoz pedig végpont mínusz kezdőpont vektort, mert ezek felelnek meg a két külön fizikai definíciónak.

**Minimális kódrecept:** A következő kódcella egy rakétapályán kiszámítja a teljes utat és a kezdőponttól végpontig vett elmozdulás nagyságát.

**Gyakori hiba:** A teljes utat és az elmozdulást ugyanannak veszik, vagy az origótól mért végső távolságot használják akkor is, ha a kezdőpont nem az origó.

**Feladatok:** Gy/8b, Gy/14b, 25_05_21/Gy8b.

```python
def egyszeru_raketa():
    m0 = 0.40
    g = 9.81

    def tomeg(t):
        return max(0.10, m0 - 0.20*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*0.2*np.pi*0.03**2*1.3*v*sebesseg
            if t <= 1.5:
                F += 7.0*sebesseg/v
        return F

    return ero, tomeg

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

ero, tomeg = egyszeru_raketa()
szog = np.radians(85.0)
raketa = num_dinam(2)
raketa.set_time_param(0.0, 40.0, 0.02)
raketa.set_mass_fun(tomeg)
raketa.set_F_fun(ero)
raketa.set_r0_v0(np.array([0.0, 1.0]), np.array([np.cos(szog), np.sin(szog)]))
raketa.set_stop_cond(talajt_ert)
raketa.full_dinam_calc()

teljes_ut = raketa.pathlength[-1]
dx = raketa.r_x[-1] - raketa.r_x[0]
dy = raketa.r_y[-1] - raketa.r_y[0]
elmozdulas = np.sqrt(dx**2 + dy**2)

print(f'Teljes út: {teljes_ut:.2f} m')
print(f'Elmozdulás nagysága: {elmozdulas:.2f} m')
```

### 1.23 Kilövési szög függvényében elért maximális sebesség

**Vizsgaszöveg:** Ábrázolja grafikonon a kilövési szög függvényében a repülés közben elért maximális sebesség nagyságát! / Milyen kilövési szög mellett lenne maximális az előző pontbeli sebesség és mennyi ennek értéke?

**Fizikai jelentés:** A kilövési szög megváltoztatja, hogyan oszlik meg a kezdősebesség vízszintes és függőleges komponensre, ez pedig a rakéta tolóerejével és közegellenállásával együtt más sebességmaximumot adhat. A kód minden szögre ugyanazt a fizikai modellt futtatja, majd a repülés közbeni `v_abs` maximumát menti el. A legnagyobb eredményhez tartozó szög adja az optimumot.

**Minimális kódrecept:** A következő kódcella szögsöpréssel ábrázolja a maximális sebességet, és kiírja a maximáló szöget.

**Gyakori hiba:** A becsapódási sebességet veszik maximumsebességnek, pedig a feladat a repülés közben előforduló legnagyobb sebességnagyságot kéri.

**Feladatok:** Gy/8c, Gy/8d, Gy/14c, Gy/14d, 25_05_21/Gy8c, 25_05_21/Gy8d.

```python
def raketa_modell():
    szaraz_tomeg = 0.10
    hajtoanyag_tomeg = 0.30
    kiaramlasi_seb = 35.0
    tomegaram = 0.20
    C = 0.20
    A = np.pi*0.03**2
    rho = 1.30
    g = 9.81

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
            if t <= hajtoanyag_tomeg/tomegaram:
                F += tomegaram*kiaramlasi_seb*sebesseg/v
        return F

    return ero, tomeg

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

def raketa_szimulacio(szog_fok):
    ero, tomeg = raketa_modell()
    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, 40.0, 0.03)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(np.array([0.0, 1.0]), np.array([np.cos(szog), np.sin(szog)]))
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

szogek = np.linspace(5.0, 85.0, 41)
max_sebessegek = []
for szog in szogek:
    raketa = raketa_szimulacio(szog)
    repulesben = raketa.r_y >= 0.0
    max_sebessegek.append(raketa.v_abs[repulesben].max())
max_sebessegek = np.array(max_sebessegek)

i_max = max_sebessegek.argmax()
print(f'Optimális szög: {szogek[i_max]:.1f} fok')
print(f'Legnagyobb repülés közbeni sebesség: {max_sebessegek[i_max]:.2f} m/s')

plt.figure(figsize=(8, 4))
plt.plot(szogek, max_sebessegek)
plt.scatter(szogek[i_max], max_sebessegek[i_max], color='red', zorder=3)
plt.xlabel('kilövési szög (fok)')
plt.ylabel('maximális sebességnagyság (m/s)')
plt.grid()
plt.show()
```

### 1.24 Épület lehetséges helye magas pályaszakasz alatt

**Vizsgaszöveg:** Egy olyan épület felett szeretnénk átlőni, melynek magassága 20 m és a lövés irányában 40 m hosszú. Milyen messze lehet a kilövés helye az épülettől, ha a kilövési szög az eredeti 85 fok?

**Fizikai jelentés:** Nem elég, hogy a rakéta valahol 20 m fölé menjen: az épület teljes 40 m-es hossza alatt a pályának végig az épület magassága fölött kell lennie. A `x_ranges_above_height` megadja azokat az x-intervallumokat, ahol a pálya egy adott magasság felett halad. A kód azért az intervallum hosszából vonja le az épület hosszát, mert az épület eleje csak úgy helyezhető el, hogy az egész épület a magas pályaszakasz alá essen.

**Minimális kódrecept:** A következő kódcella megkeresi a 20 m feletti x-intervallumokat, majd kiszámolja, hova tehető egy 40 m hosszú épület.

**Gyakori hiba:** Csak egyetlen pontban ellenőrzik a magasságot, pedig az épület teljes hosszán teljesülnie kell a feltételnek.

**Feladatok:** Gy/8e, 25_05_21/Gy8e.

```python
def egyszeru_magas_raketa():
    def tomeg(t):
        return 0.40

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*9.81], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.02*v*sebesseg
            if t < 1.2:
                F += 18.0*sebesseg/v
        return F

    return ero, tomeg

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

ero, tomeg = egyszeru_magas_raketa()
szog = np.radians(85.0)
raketa = num_dinam(2)
raketa.set_time_param(0.0, 30.0, 0.02)
raketa.set_mass_fun(tomeg)
raketa.set_F_fun(ero)
raketa.set_r0_v0(np.array([0.0, 1.0]), np.array([2.0*np.cos(szog), 2.0*np.sin(szog)]))
raketa.set_stop_cond(talajt_ert)
raketa.full_dinam_calc()

magassag = 20.0
epulet_hossz = 40.0
intervallumok = raketa.x_ranges_above_height(magassag)

for x1, x2 in intervallumok:
    if x2 - x1 >= epulet_hossz:
        print(f'Az épület eleje {x1:.2f} m és {x2 - epulet_hossz:.2f} m közé tehető.')
```

### 1.25 Első talajérés ideje, helye és sebessége

**Vizsgaszöveg:** Mikor, az eldobás helyétől milyen távol és milyen sebességgel éri el a talajt először a labda?

**Fizikai jelentés:** Az első talajérés az első olyan esemény, amikor a labda magassága eléri a talajszintet. A becsapódási távolság a vízszintes koordináta, a becsapódási sebesség pedig a sebességvektor nagysága ugyanebben az időpontban. A kód azért az első `r_y = 0` átlépés indexét használja, mert ez jelöli a mozgás első érintkezését a talajjal.

**Minimális kódrecept:** A következő kódcella egy dobott labda pályáján megkeresi az első talajátlépést, majd kiírja az időt, vízszintes távolságot és sebességnagyságot.

**Gyakori hiba:** Az utolsó pályapontot veszik első talajérésnek, pedig pattogó vagy tovább szimulált mozgásnál később több talajérintés is lehet.

**Feladatok:** Gy/9b, F2-2026/b.

```python
def labda_levegoben():
    m = 0.59
    A = np.pi*(0.76/(2*np.pi))**2
    C = 0.47
    rho = 1.30
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    return ero, tomeg

def talajt_ert(hely, sebesseg):
    return hely[1] < 0.0

ero, tomeg = labda_levegoben()
szog = np.radians(40.0)
labda = num_dinam(2)
labda.set_time_param(0.0, 10.0, 0.002)
labda.set_mass_fun(tomeg)
labda.set_F_fun(ero)
labda.set_r0_v0(np.array([0.0, 1.9]), np.array([10.0*np.cos(szog), 10.0*np.sin(szog)]))
labda.set_stop_cond(talajt_ert)
labda.full_dinam_calc()

print(f'Első talajérés ideje: {labda.t[-1]:.2f} s')
print(f'Talajérés távolsága: {labda.r_x[-1]:.2f} m')
print(f'Talajérési sebesség: {labda.v_abs[-1]:.2f} m/s')
```

### 1.26 Átrepül-e a labda adott akadály felett?

**Vizsgaszöveg:** Ha tőlünk 3 m-re áll valaki, aki ugrással együtt 2,6 m magasra tudja nyújtani a kezét, akkor átrepül-e felette a labda?

**Fizikai jelentés:** Az akadály feletti átrepülésnél nem a pálya legnagyobb magassága számít, hanem a labda magassága éppen az akadály vízszintes helyén. Ha az `x = 3 m` helyen a pálya `y` értéke nagyobb 2,6 m-nél, akkor a labda átrepül felette. A kód ezt a geometriai összehasonlítást végzi el a pályaadatokból.

**Minimális kódrecept:** A következő kódcella `flies_over_obstacle` segítségével eldönti, hogy a labda átmegy-e az adott személy fölött.

**Gyakori hiba:** A maximális magasságot hasonlítják az akadályhoz, pedig lehet, hogy a labda máshol magas, nem az akadály helyén.

**Feladatok:** Gy/9c, F2-2026/c.

```python
t = np.arange(0.0, 2.0, 0.002)
g = 9.81
v0 = 10.0
szog = np.radians(40.0)
x = v0*np.cos(szog)*t
y = 1.9 + v0*np.sin(szog)*t - 0.5*g*t**2

labda = num_kinem(2)
labda.t = t
labda.r = np.column_stack((x, y))
labda.full_kinem_calc()

atrepul, y_ertekek = labda.flies_over_obstacle(3.0, 2.6)
print(f'Átrepül a labda a személy felett? {atrepul}')
if len(y_ertekek) > 0:
    print(f'A labda magassága x = 3 m-nél kb. {max(y_ertekek):.2f} m')
```

```python
# Gy/10d kiegészítés: nem azt kérdezzük, átrepül-e a fal felett,
# hanem azt, milyen magasan találja el az x = 8 m helyen lévő falat.
def teniszlabda_palya():
    m = 0.058
    A = np.pi*(0.067/2)**2
    C = 0.55
    rho = 1.225
    g = 9.81
    v0 = 15.0
    szog = np.radians(70.0)

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    labda = num_dinam(2)
    labda.set_time_param(0.0, 8.0, 0.004)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.8]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

labda = teniszlabda_palya()
y_falnal = labda.y_values_at_x(8.0)
print(f'A labda a 8 m-re lévő falat kb. {y_falnal[0]:.2f} m magasan éri el.')
```

### 1.27 Lokálisan legmagasabb pontok táblázata

**Vizsgaszöveg:** Készítsen táblázatot, mely a pálya lokálisan legmagasabb pontjaira adja meg sorra az időpontot, x és y koordinátát és a sebesség nagyságát!

**Fizikai jelentés:** Egy dobási ív tetején a függőleges sebesség pozitívból negatívba vált: előtte a test emelkedik, utána süllyed. Ezért a lokális csúcspontokat a `v_y` előjelváltásai alapján lehet megtalálni. A kód ezeknél az indexeknél ugyanabból az időpillanatból olvassa ki az időt, helyet és sebességnagyságot.

**Minimális kódrecept:** A következő kódcella egy pattogó labda több ívének felső pontjait táblázatszerűen kiírja.

**Gyakori hiba:** Csak a globális maximumot keresik meg, pedig a pattogó mozgásban minden ívnek külön lokális legmagasabb pontja van.

**Feladatok:** Gy/9d, F2-2026/d, F2-2026/e.

```python
def pattogo_labda_modell():
    m = 0.59
    A = np.pi*(0.76/(2*np.pi))**2
    C = 0.47
    rho = 1.30
    D = 4000.0
    F_f = 3.0
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        if hely[1] < 0.0:
            F[1] += D*(-hely[1])
            if abs(sebesseg[0]) > 1e-10:
                F[0] -= F_f*np.sign(sebesseg[0])
        return F

    return ero, tomeg

ero, tomeg = pattogo_labda_modell()
szog = np.radians(40.0)
labda = num_dinam(2)
labda.set_time_param(0.0, 12.0, 0.003)
labda.set_mass_fun(tomeg)
labda.set_F_fun(ero)
labda.set_r0_v0(np.array([0.0, 1.9]), np.array([10.0*np.cos(szog), 10.0*np.sin(szog)]))
labda.set_stop_cond(None)
labda.full_dinam_calc()

csucsok = np.where((labda.v_y[:-1] > 0.0) & (labda.v_y[1:] <= 0.0))[0] + 1

print('sorszám\tidő (s)\tx (m)\ty (m)\tsebesség (m/s)')
for sorszam, i in enumerate(csucsok[:8], start=1):
    print(f'{sorszam}\t{labda.t[i]:.2f}\t{labda.r_x[i]:.2f}\t{labda.r_y[i]:.2f}\t{labda.v_abs[i]:.2f}')
```

### 1.28 Első lepattanás távolságának optimalizálása eldobási szöggel

**Vizsgaszöveg:** Ha csupán az eldobás szögét változtatja, milyen szög esetén lesz maximális az első lepattanás távolsága az eldobás helyétől mérve és mekkora ez a távolság?

**Fizikai jelentés:** Az eldobási szög egyszerre befolyásolja a vízszintes sebességet és a levegőben töltött időt. Az első lepattanás távolsága ott maximális, ahol ez a két hatás együtt a legnagyobb vízszintes elmozdulást adja az első talajérésig. A kód ezért sok szöget kipróbál, minden szögnél az első talajérés `r_x[-1]` értékét méri, majd ezek maximumát keresi.

**Minimális kódrecept:** A következő kódcella szögsöpréssel megkeresi az első lepattanás legnagyobb távolságát.

**Gyakori hiba:** A teljes 20 másodperces pattogó pálya végpontját optimalizálják, pedig a feladat csak az első lepattanás távolságát kéri.

**Feladatok:** Gy/9e, F2-2026/f.

```python
def elso_talajeres_tavolsag(szog_fok):
    m = 0.59
    A = np.pi*(0.76/(2*np.pi))**2
    C = 0.47
    rho = 1.30
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    labda = num_dinam(2)
    labda.set_time_param(0.0, 8.0, 0.004)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.9]), np.array([10.0*np.cos(szog), 10.0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda.r_x[-1]

szogek = np.linspace(5.0, 85.0, 81)
tavolsagok = np.array([elso_talajeres_tavolsag(szog) for szog in szogek])
i_max = tavolsagok.argmax()

print(f'Optimális szög: {szogek[i_max]:.1f} fok')
print(f'Első lepattanás maximális távolsága: {tavolsagok[i_max]:.2f} m')

plt.figure(figsize=(8, 4))
plt.plot(szogek, tavolsagok)
plt.scatter(szogek[i_max], tavolsagok[i_max], color='red', zorder=3)
plt.xlabel('eldobási szög (fok)')
plt.ylabel('első lepattanás távolsága (m)')
plt.grid()
plt.show()
```

### 1.29 Gyorsulás nagyságának maximuma

**Vizsgaszöveg:** Mikor és hol lesz a labda gyorsulásának nagysága a legnagyobb, és mennyi ennek a gyorsulásnak az értéke?

**Fizikai jelentés:** Közegellenállás mellett a gyorsulás nem csak a gravitációból áll, mert a légellenállás is erőt fejt ki a testre. A gyorsulás nagyságának maximuma ott van, ahol az eredő erő nagysága a legnagyobb. A kód azért az `a_abs.argmax()` indexet használja, mert `a_abs` a teljes gyorsulásvektor nagyságát adja meg minden időpontra.

**Minimális kódrecept:** A következő kódcella meghatározza a gyorsulásmaximum időpontját, helyét és értékét.

**Gyakori hiba:** Csak a függőleges gyorsulást vagy csak `g` értékét nézik, pedig közegellenállással az eredő gyorsulás nagysága ettől eltérhet.

**Feladatok:** Gy/10b.

```python
def teniszlabda_palya():
    m = 0.058
    A = np.pi*(0.067/2)**2
    C = 0.55
    rho = 1.225
    g = 9.81
    v0 = 15.0
    szog = np.radians(70.0)

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    labda = num_dinam(2)
    labda.set_time_param(0.0, 8.0, 0.004)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.8]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

labda = teniszlabda_palya()
i = labda.a_abs.argmax()
print(f'Időpont: {labda.t[i]:.3f} s')
print(f'Hely: x = {labda.r_x[i]:.2f} m, y = {labda.r_y[i]:.2f} m')
print(f'Legnagyobb gyorsulás: {labda.a_abs[i]:.2f} m/s²')
```

### 1.30 Mikor lesz egy mennyiség adott értékű?

**Vizsgaszöveg:** Mikor és hol lesz a labda gyorsulásának nagysága épp g?

**Fizikai jelentés:** Ez küszöbérték-elérési kérdés: azt keressük, mikor veszi fel a gyorsulás nagysága a megadott fizikai értéket. Táblázatos numerikus adatoknál ritka a pontos egyezés, ezért az átlépést kell keresni két szomszédos minta között. Az `arg_eq` éppen ezt a fizikai eseményt közelíti: megadja, hol lépi át a mennyiség a keresett értéket.

**Minimális kódrecept:** A következő kódcella megkeresi, mikor lesz `a_abs` közel `g` értékű, és kiolvassa az időt és helyet.

**Gyakori hiba:** Pontos `== g` összehasonlítást használnak lebegőpontos adatokon, ami gyakran nem talál semmit.

**Feladatok:** Gy/10c, Vizsga-minta-01/F3b.

```python
def teniszlabda_palya():
    m = 0.058
    A = np.pi*(0.067/2)**2
    C = 0.55
    rho = 1.225
    g = 9.81
    v0 = 15.0
    szog = np.radians(70.0)

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    labda = num_dinam(2)
    labda.set_time_param(0.0, 8.0, 0.004)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.8]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

labda = teniszlabda_palya()
g = 9.81
indexek = arg_eq(labda.a_abs, g)
i = indexek[0]
print(f'a_abs = g közelében: t = {labda.t[i]:.3f} s')
print(f'Hely: x = {labda.r_x[i]:.2f} m, y = {labda.r_y[i]:.2f} m')
print(f'Gyorsulás: {labda.a_abs[i]:.2f} m/s²')
```

```python
# Vizsga-minta-01/F3b kiegészítés: nem gyorsulásértéket, hanem adott
# vízszintes távolság első elérésének idejét keressük.
t = np.linspace(0.0, 8.0, 801)
x = 9.0*t*(1.0 - np.exp(-t/1.5))
y = 1.9 + 4.0*np.sin(np.radians(50))*t - 0.5*9.81*t**2
y = np.maximum(y, 0.0)

labda = num_kinem(2)
labda.t = t
labda.r = np.column_stack([x, y])
labda.full_kinem_calc()

cel_x = 20.0
i = np.argmax(labda.r_x >= cel_x)
print(f'A labda {labda.t[i]:.2f} s után jut el vízszintesen {cel_x:.0f} m-re.')
print(f'Ekkori magasság: {labda.r_y[i]:.2f} m')
```

### 1.33 Függőleges lövés: legnagyobb magasság és visszaesés

```python
def lovedek_fuggoleges():
    m = 0.0075
    A = np.pi*(0.00901/2)**2
    C = 0.32
    rho = 1.3
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    lovedek = num_dinam(2)
    lovedek.set_time_param(0.0, 80.0, 0.002)
    lovedek.set_mass_fun(tomeg)
    lovedek.set_F_fun(ero)
    lovedek.set_r0_v0(np.array([0.0, 3.0]), np.array([0.0, 380.0]))
    lovedek.set_stop_cond(talajt_ert)
    lovedek.full_dinam_calc()
    return lovedek

lovedek = lovedek_fuggoleges()
i_csucs = lovedek.r_y.argmax()
print(f'Legnagyobb magasság: {lovedek.r_y[i_csucs]:.2f} m')
print(f'Csúcspont ideje: {lovedek.t[i_csucs]:.2f} s')
print(f'Visszaesés ideje: {lovedek.t[-1]:.2f} s')
print(f'Visszaesési sebesség: {lovedek.v_abs[-1]:.2f} m/s')
```

### 1.34 Kilövési szög szerint két eredmény: lőtávolság és becsapódási sebesség

**Vizsgaszöveg:** Változtassuk a kilövés vízszintessel bezárt szögét 0 és 90 fok közt. Ábrázolja a kilövés szögének függvényében a lövés távolságát és a becsapódás sebességét!

**Fizikai jelentés:** Ugyanaz a bemenő paraméter, a kilövési szög, egyszerre befolyásolja a repülési időt, a lőtávolságot és a becsapódási sebességet. A kód minden szögre lefuttatja ugyanazt a dinamikai modellt, majd két külön kimenetet ment: `r_x[-1]` és `v_abs[-1]`. Ez azért válaszol a kérdésre, mert mindkét mennyiség ugyanahhoz a becsapódási eseményhez tartozik.

**Minimális kódrecept:** A következő kódcella szögsöpréssel két grafikont készít: lőtávolság és becsapódási sebesség a szög függvényében.

**Gyakori hiba:** Csak a lőtávolságot ábrázolják, és kihagyják a becsapódási sebesség második grafikonját.

**Feladatok:** Gy/11e, Gy/11f.

```python
def lovedek_szog(szog_fok):
    m = 0.0075
    A = np.pi*(0.00901/2)**2
    C = 0.32
    rho = 1.3
    g = 9.81
    v0 = 380.0

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    lovedek = num_dinam(2)
    lovedek.set_time_param(0.0, 80.0, 0.01)
    lovedek.set_mass_fun(tomeg)
    lovedek.set_F_fun(ero)
    lovedek.set_r0_v0(np.array([0.0, 3.0]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    lovedek.set_stop_cond(talajt_ert)
    lovedek.full_dinam_calc()
    return lovedek

szogek = np.linspace(0.0, 85.0, 36)
lotav = []
becsap_seb = []
for szog in szogek:
    lovedek = lovedek_szog(szog)
    lotav.append(lovedek.r_x[-1])
    becsap_seb.append(lovedek.v_abs[-1])
lotav = np.array(lotav)
becsap_seb = np.array(becsap_seb)

i_max = lotav.argmax()
print(f'Maximális lőtávolság szöge: {szogek[i_max]:.1f} fok')
print(f'Maximális lőtávolság: {lotav[i_max]:.2f} m')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(szogek, lotav)
ax1.scatter(szogek[i_max], lotav[i_max], color='red')
ax1.set_xlabel('kilövési szög (fok)')
ax1.set_ylabel('lőtávolság (m)')
ax1.grid()

ax2.plot(szogek, becsap_seb)
ax2.set_xlabel('kilövési szög (fok)')
ax2.set_ylabel('becsapódási sebesség (m/s)')
ax2.grid()
plt.show()
```

### 1.35 Hely-, sebesség- és gyorsulás-idő függvény talajra érésig

**Vizsgaszöveg:** Ábrázolja egy pingpong-labda hely-, sebesség- és gyorsulás-idő függvényét a talajra érésig, ha egy lépcsőházban 25 m magasról kezdősebesség nélkül leejtjük!

**Fizikai jelentés:** Szabadesésnél és közegellenállásnál a hely, sebesség és gyorsulás együtt írja le a mozgásállapot változását. A talajra érésig tartó ábrázolás azért fontos, mert utána már más fizikai kölcsönhatás, a talajerő is belépne. A kód stop feltétellel a talajig szimulál, majd ugyanazon időtengelyen rajzolja a `r_y`, `v_abs` és `a_abs` mennyiségeket.

**Minimális kódrecept:** A következő kódcella három panelen ábrázolja a helyet, sebességnagyságot és gyorsulásnagyságot a talajra érésig.

**Gyakori hiba:** A labdát tovább szimulálják talajerővel, pedig ez a kérdés csak a talajra érésig tartó esést kéri.

**Feladatok:** Gy/12a.

```python
def pingpong_esik():
    m = 0.0027
    A = np.pi*(0.040/2)**2
    C = 0.47
    rho = 1.3
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    labda = num_dinam(2)
    labda.set_time_param(0.0, 20.0, 0.002)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 25.0]), np.array([0.0, 0.0]))
    labda.set_stop_cond(talajt_ert)
    labda.full_dinam_calc()
    return labda

labda = pingpong_esik()
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 8), sharex=True)
ax1.plot(labda.t, labda.r_y)
ax1.set_ylabel('y (m)')
ax1.grid()
ax2.plot(labda.t, labda.v_abs)
ax2.set_ylabel('sebesség (m/s)')
ax2.grid()
ax3.plot(labda.t, labda.a_abs)
ax3.set_xlabel('idő (s)')
ax3.set_ylabel('gyorsulás (m/s²)')
ax3.grid()
plt.show()
```

### 1.36 Földet érési helyek táblázata pattogó mozgásnál

**Vizsgaszöveg:** Határozza meg a földet érési helyek időpontjait és egy kis táblázatban print() parancsokkal adja meg ezek időpontját, x koordinátáját és a földnek csapódás sebességét!

**Fizikai jelentés:** Pattogó mozgásnál a labda többször is eléri a talajt, ezért nem elég az első becsapódást megkeresni. Egy földet érés akkor történik, amikor a magasság a talajszintet eléri, és a függőleges sebesség lefelé mutat. A kód ezért az `r_y = 0` átlépéseket szűri lefelé haladásra, majd minden eseményhez kiolvassa az időt, x koordinátát és `v_abs` értéket.

**Minimális kódrecept:** A következő kódcella pattogó labdára táblázatot ír a földet érési eseményekről.

**Gyakori hiba:** Minden talajszint-átlépést felsorolnak, beleértve a felfelé pattanás utáni áthaladást is, pedig csak a földnek csapódások kellenek.

**Feladatok:** Gy/12c.

```python
def pattogo_pingpong():
    m = 0.0027
    A = np.pi*(0.040/2)**2
    C = 0.47
    rho = 1.3
    D = 1000.0
    F_f = 0.2
    g = 9.81

    def tomeg(t):
        return m

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5*C*A*rho*v*sebesseg
        if hely[1] < 0.0:
            F[1] += D*(-hely[1])
            if abs(sebesseg[0]) > 1e-10:
                F[0] -= F_f*np.sign(sebesseg[0])
        return F

    szog = np.radians(30.0)
    v0 = 40.0/3.6
    labda = num_dinam(2)
    labda.set_time_param(0.0, 12.0, 0.003)
    labda.set_mass_fun(tomeg)
    labda.set_F_fun(ero)
    labda.set_r0_v0(np.array([0.0, 1.1]), np.array([v0*np.cos(szog), v0*np.sin(szog)]))
    labda.set_stop_cond(None)
    labda.full_dinam_calc()
    return labda

labda = pattogo_pingpong()
jeloltek = arg_eq(labda.r_y, 0.0)
foldet_eres = [i for i in jeloltek if labda.v_y[i] < 0.0]

szurt = []
for i in foldet_eres:
    if not szurt or labda.t[i] - labda.t[szurt[-1]] > 0.08:
        szurt.append(i)

print('idő (s)\tx (m)\tbecsapódási sebesség (m/s)')
for i in szurt[:8]:
    print(f'{labda.t[i]:.2f}\t{labda.r_x[i]:.2f}\t{labda.v_abs[i]:.2f}')
```

### 1.37 Sebességnagyság csökkenő szakaszainak jelölése

**Vizsgaszöveg:** Rajzolja fel a rakéta pályáját a földbe csapódásig! Ezen jelölje zölddel azokat a részeket, ahol a rakéta sebességének nagysága csökkent!

**Fizikai jelentés:** A sebességnagyság akkor csökken, amikor a gyorsulás pálya menti komponense a pillanatnyi sebességgel ellentétes irányú. Ezért a `a_t_abs < 0` feltétel fizikailag azt jelzi, hogy az eredő erőnek van lassító komponense. A pályán pontosan ezeket a pontokat színezzük ki, mert ott történik a tempó csökkenése, nem feltétlenül ott, ahol a gyorsulás nagysága nagy.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Sokan `a_abs < 0` feltételt próbálnak írni, de a gyorsulás nagysága nem lehet negatív; a lassulást az előjeles tangenciális gyorsulás mutatja.

**Feladatok:** Gy/13a.


```python
def raketa_szimulacio(hajtoanyag_tomeg=0.35, szaraz_tomeg=0.15,
                     v_kiaramlas=45.0, tomegaram=0.2, sugar=0.04,
                     alaktenyezo=0.2, szog_fok=85.0, kezdo_v=1.0,
                     kezdo_y=1.0, dt=0.02, t_max=25.0):
    rho = 1.2
    g = 9.81
    A = np.pi * sugar**2
    egesi_ido = hajtoanyag_tomeg / tomegaram

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5 * alaktenyezo * A * rho * v * sebesseg
            if t <= egesi_ido:
                F += tomegaram * v_kiaramlas * sebesseg / v
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, t_max, dt)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(
        np.array([0.0, kezdo_y]),
        np.array([kezdo_v*np.cos(szog), kezdo_v*np.sin(szog)])
    )
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

raketa = raketa_szimulacio(
    hajtoanyag_tomeg=5.0, szaraz_tomeg=1.5, v_kiaramlas=120.0,
    tomegaram=1.4, sugar=0.06, alaktenyezo=0.2,
    szog_fok=85.0, kezdo_v=1.0, kezdo_y=1.0,
    dt=0.02, t_max=45.0
)

lassul = raketa.a_t_abs < 0

plt.figure(figsize=(7, 5))
plt.plot(raketa.r_x, raketa.r_y, color='0.75', label='rakétapálya')
plt.scatter(raketa.r_x[lassul], raketa.r_y[lassul], s=10, color='green',
            label='csökken a sebességnagyság')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

### 1.38 Mozgási energia ábrázolása változó tömegű rakétánál

**Vizsgaszöveg:** Ábrázolja a rakéta mozgási energiáját az idő függvényében!

**Fizikai jelentés:** Rakétánál a mozgási energia nem csak a sebességnagyságtól, hanem az időben változó tömegtől is függ. A fizikai képlet továbbra is E = 1/2 m v², de itt az `m` minden időpillanatban más lehet, mert fogy a hajtóanyag. Ezért működik a `kinetic_energy(raketa.m)` recept: ugyanahhoz az időponthoz tartozó tömeget és sebességnagyságot szorozza össze.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba az üres tömeggel vagy a kezdőtömeggel végig állandó energiát számolni, pedig a rakéta tömege repülés közben csökken.

**Feladatok:** Gy/13b.


```python
def raketa_szimulacio(hajtoanyag_tomeg=0.35, szaraz_tomeg=0.15,
                     v_kiaramlas=45.0, tomegaram=0.2, sugar=0.04,
                     alaktenyezo=0.2, szog_fok=85.0, kezdo_v=1.0,
                     kezdo_y=1.0, dt=0.02, t_max=25.0):
    rho = 1.2
    g = 9.81
    A = np.pi * sugar**2
    egesi_ido = hajtoanyag_tomeg / tomegaram

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5 * alaktenyezo * A * rho * v * sebesseg
            if t <= egesi_ido:
                F += tomegaram * v_kiaramlas * sebesseg / v
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, t_max, dt)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(
        np.array([0.0, kezdo_y]),
        np.array([kezdo_v*np.cos(szog), kezdo_v*np.sin(szog)])
    )
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

raketa = raketa_szimulacio(
    hajtoanyag_tomeg=5.0, szaraz_tomeg=1.5, v_kiaramlas=120.0,
    tomegaram=1.4, sugar=0.06, alaktenyezo=0.2,
    szog_fok=85.0, kezdo_v=1.0, kezdo_y=1.0,
    dt=0.02, t_max=45.0
)

E_mozgasi = raketa.kinetic_energy(raketa.m)

plt.figure(figsize=(8, 4))
plt.plot(raketa.t, E_mozgasi)
plt.xlabel('idő (s)')
plt.ylabel('mozgási energia (J)')
plt.grid()
plt.show()
```

### 1.39 Idő és megtett út gyorsulási küszöb felett

**Vizsgaszöveg:** Határozza meg, mennyi ideig volt a rakéta gyorsulásának abszolút értéke 5g-nél nagyobb, és ez alatt mekkora távolságot tett meg a rakéta!

**Fizikai jelentés:** A kérdés nem a sebességre, hanem az eredő gyorsulás nagyságára tesz küszöbfeltételt. A `a_abs > 5*g` maszk ezért pontosan azokat az időlépéseket választja ki, amikor a testre ható eredő hatás 5g-nél erősebb gyorsulást okoz. Az időt a kijelölt időlépésekből, az utat pedig ugyanazon maszk alatt a `pathlength` növekményeiből összegezve kapjuk, mert a pályahossz-darabok adják a tényleges megtett távolságot.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Tipikus hiba, hogy a teljes lőtávolságot vagy `r_x[-1]`-et adják meg, pedig a kérdés csak a nagy gyorsulású szakasz alatt megtett pályahosszt kéri.

**Feladatok:** Gy/13c.


```python
def raketa_szimulacio(hajtoanyag_tomeg=0.35, szaraz_tomeg=0.15,
                     v_kiaramlas=45.0, tomegaram=0.2, sugar=0.04,
                     alaktenyezo=0.2, szog_fok=85.0, kezdo_v=1.0,
                     kezdo_y=1.0, dt=0.02, t_max=25.0):
    rho = 1.2
    g = 9.81
    A = np.pi * sugar**2
    egesi_ido = hajtoanyag_tomeg / tomegaram

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5 * alaktenyezo * A * rho * v * sebesseg
            if t <= egesi_ido:
                F += tomegaram * v_kiaramlas * sebesseg / v
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, t_max, dt)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(
        np.array([0.0, kezdo_y]),
        np.array([kezdo_v*np.cos(szog), kezdo_v*np.sin(szog)])
    )
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

raketa = raketa_szimulacio(
    hajtoanyag_tomeg=5.0, szaraz_tomeg=1.5, v_kiaramlas=120.0,
    tomegaram=1.4, sugar=0.06, alaktenyezo=0.2,
    szog_fok=85.0, kezdo_v=1.0, kezdo_y=1.0,
    dt=0.02, t_max=45.0
)

g = 9.81
feltetel = raketa.a_abs > 5*g
utdarabok = np.diff(raketa.pathlength, prepend=0.0)

ido = feltetel.sum() * raketa.delta_t
tavolsag = utdarabok[feltetel].sum()

print(f'5g feletti gyorsulás ideje: {ido:.3f} s')
print(f'Ez alatt megtett út: {tavolsag:.2f} m')
```

### 1.40 Hajtóanyag-tömeg optimalizálása lőtávolságra

**Vizsgaszöveg:** Határozza meg, milyen hajtóanyag-tömeg mellett lesz a lövés távolsága sík terepen maximális!

**Fizikai jelentés:** A hajtóanyag növelése egyszerre ad több tolóerő-időtartamot és nagyobb induló tömeget, ezért nem nyilvánvaló, hogy a legtöbb hajtóanyag adja a legnagyobb távolságot. A kód azért működik fizikailag, mert minden hajtóanyag-tömegre ugyanazt a mozgásegyenletet futtatja, majd a becsapódási x koordinátát, vagyis a lőtávolságot hasonlítja össze. Az optimum az a paraméterérték, amelynél ez a fizikai kimenet a legnagyobb.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba a pályahosszat optimalizálni a vízszintes lőtávolság helyett; sík terepen a lőtávolság a becsapódási `r_x[-1]`.

**Feladatok:** Gy/13d.


```python
def raketa_szimulacio(hajtoanyag_tomeg=0.35, szaraz_tomeg=0.15,
                     v_kiaramlas=45.0, tomegaram=0.2, sugar=0.04,
                     alaktenyezo=0.2, szog_fok=85.0, kezdo_v=1.0,
                     kezdo_y=1.0, dt=0.02, t_max=25.0):
    rho = 1.2
    g = 9.81
    A = np.pi * sugar**2
    egesi_ido = hajtoanyag_tomeg / tomegaram

    def tomeg(t):
        return max(szaraz_tomeg, szaraz_tomeg + hajtoanyag_tomeg - tomegaram*t)

    def ero(t, hely, sebesseg, m):
        F = np.array([0.0, -m*g], np.float64)
        v = np.sqrt((sebesseg**2).sum())
        if v > 1e-10:
            F -= 0.5 * alaktenyezo * A * rho * v * sebesseg
            if t <= egesi_ido:
                F += tomegaram * v_kiaramlas * sebesseg / v
        return F

    def talajt_ert(hely, sebesseg):
        return hely[1] < 0.0

    szog = np.radians(szog_fok)
    raketa = num_dinam(2)
    raketa.set_time_param(0.0, t_max, dt)
    raketa.set_mass_fun(tomeg)
    raketa.set_F_fun(ero)
    raketa.set_r0_v0(
        np.array([0.0, kezdo_y]),
        np.array([kezdo_v*np.cos(szog), kezdo_v*np.sin(szog)])
    )
    raketa.set_stop_cond(talajt_ert)
    raketa.full_dinam_calc()
    return raketa

hajtoanyagok = np.linspace(1.0, 8.0, 22)
lotavolsagok = []

for mh in hajtoanyagok:
    raketa = raketa_szimulacio(
        hajtoanyag_tomeg=mh, szaraz_tomeg=1.5, v_kiaramlas=120.0,
        tomegaram=1.4, sugar=0.06, alaktenyezo=0.2,
        szog_fok=85.0, kezdo_v=1.0, kezdo_y=1.0,
        dt=0.04, t_max=45.0
    )
    lotavolsagok.append(raketa.r_x[-1])

lotavolsagok = np.array(lotavolsagok)
i = lotavolsagok.argmax()

print(f'Optimális hajtóanyag-tömeg: {hajtoanyagok[i]:.2f} kg')
print(f'Maximális lőtávolság: {lotavolsagok[i]:.2f} m')

plt.figure(figsize=(8, 4))
plt.plot(hajtoanyagok, lotavolsagok)
plt.scatter(hajtoanyagok[i], lotavolsagok[i], color='red', zorder=3)
plt.xlabel('hajtóanyag-tömeg (kg)')
plt.ylabel('lőtávolság (m)')
plt.grid()
plt.show()
```

### 1.41 Teljes megtett távolság és átlagos sebességnagyság

**Vizsgaszöveg:** Határozza meg a teljes útra a megtett távolságot és az átlagos sebességnagyságot!

**Fizikai jelentés:** A megtett távolság a pálya mentén összegzett ívhossz, ezért GPS-szerű adatsornál nem a végpontok távolságából számoljuk. Az átlagos sebességnagyság fizikailag a teljes megtett út osztva a teljes eltelt idővel. A `pathlength[-1]` és a teljes időtartam hányadosa ezért pontosan azt méri, milyen átlagos tempóval járta be a test a tényleges útvonalat.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Sokan az elmozdulás nagyságát osztják az idővel, ami kanyargós pályán nem az átlagos sebességnagyságot, hanem az átlagsebesség-vektor nagyságát adja.

**Feladatok:** Vizsga-minta-01/F2a.


```python
t = np.linspace(0.0, 120.0, 601)
sebesseg = 16.0 + 4.0*np.sin(2*np.pi*t/45.0)
irany = 0.25*np.sin(2*np.pi*t/70.0)

vx = sebesseg*np.cos(irany)
vy = sebesseg*np.sin(irany)
x = integ(t, vx)
y = integ(t, vy)

auto = num_kinem(2)
auto.t = t
auto.r = np.column_stack([x, y])
auto.full_kinem_calc()

teljes_ut = auto.pathlength[-1]
atlagsebesseg = teljes_ut / (auto.t[-1] - auto.t[0])

print(f'Megtett távolság: {teljes_ut:.1f} m')
print(f'Átlagos sebességnagyság: {atlagsebesseg:.2f} m/s')
```

### 1.42 Mikortól nem teljesül többé egy magassági feltétel?

**Vizsgaszöveg:** Mi az az időpont, ami után a labda már sosem emelkedik 1,5 m-nél magasabbra?

**Fizikai jelentés:** Pattogó mozgásnál az energiaveszteség miatt az egymást követő csúcsok egyre alacsonyabbak. A kérdés nem az első 1,5 m alá kerülést kéri, hanem az utolsó olyan időpontot, amikor a labda még 1,5 m fölött van. Ezért a kód a `r_y > 1.5` feltétel utolsó igaz pontját keresi meg, és az ezutáni időpontot adja válaszként.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba az első lefelé átlépést megadni, pedig a labda később még visszapattanhat 1,5 m fölé.

**Feladatok:** Vizsga-minta-01/F3c.


```python
t = np.linspace(0.0, 28.0, 2801)
x = 6.0 * (1.0 - np.exp(-t/9.0)) + 0.35*t
y = 0.12 + 2.1*np.exp(-0.11*t)*np.maximum(np.sin(2.6*t), 0.0)

labda = num_kinem(2)
labda.t = t
labda.r = np.column_stack([x, y])
labda.full_kinem_calc()

hatar = 1.5
magasan = labda.r_y > hatar
utolso_magas = np.where(magasan)[0][-1]
ezutan = labda.t[min(utolso_magas + 1, len(labda.t) - 1)]

print(f'{ezutan:.2f} s után a labda már nem emelkedik {hatar:.1f} m fölé.')

plt.figure(figsize=(8, 4))
plt.plot(labda.t, labda.r_y)
plt.axhline(hatar, color='red', linestyle='--', label='1,5 m')
plt.axvline(ezutan, color='green', linestyle=':', label='ettől kezdve már nem')
plt.xlabel('idő (s)')
plt.ylabel('magasság (m)')
plt.grid()
plt.legend()
plt.show()
```

### 1.43 Mérési bizonytalanság terjedése szabadesésből számolt g-re

**Vizsgaszöveg:** Egy testet h magasról leejtünk és mérjük esésének te idejét. Mekkora lesz ezekből kiszámolva g értéke és σ(g) szórása?

**Fizikai jelentés:** Szabadesésnél, elhanyagolható közegellenállás mellett, a megtett út h = 1/2 g t², ezért g = 2h/t². A bizonytalanság azért terjed a parciális deriváltakon keresztül, mert a kis mérési hibák a képlet érzékenysége szerint változtatják meg a számolt g-t. Független h és t mérésnél a relatív bizonytalanságok négyzetesen adódnak össze, a t kitevője miatt a időhiba kétszeres súllyal szerepel.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba, hogy a relatív időhibát nem szorozzák kettővel, pedig a képletben `t²` van a nevezőben.

**Feladatok:** Vizsga-minta-01/F1.


```python
h = 3.0
sigma_h = 0.01
t_e = 0.783
sigma_t = 0.002

g = 2*h / t_e**2
sigma_g = g * np.sqrt((sigma_h/h)**2 + (2*sigma_t/t_e)**2)

print(f'g = {g:.3f} m/s²')
print(f'sigma(g) = {sigma_g:.3f} m/s²')
```

A recept azért működik, mert `v_abs` a pillanatnyi tempó, `t` az idő, `pathlength` pedig a pálya menti pozíció.

### 3.2 Pályagörbe ábrázolása

### 3.3 Feltételnek megfelelő pályaszakaszok kiemelése

**Vizsgaszöveg:** Készítsen ábrát, melyen a pálya szürke, és rajta színnel be vannak jelölve a feltételnek megfelelő szakaszok!

**Fizikai jelentés:** Ilyenkor először egy fizikai feltételt alakítunk logikai maszkká, például nagy sebesség, erős fékezés vagy erős kanyarodás. A maszk ugyanazokra az időpontokra vonatkozik, mint a pályapontok, ezért a `r_x[maszk]`, `r_y[maszk]` pontok kiszínezése pontosan a megfelelő pályarészeket emeli ki.

### 3.4 Gyorsulásnagyság-idő grafikon

**Vizsgaszöveg:** Ábrázolja a gyorsulás nagyságát az idő függvényében!

**Fizikai jelentés:** A gyorsulásvektor nagysága azt mutatja, mennyire erősen változik a sebességvektor, függetlenül attól, hogy irány- vagy tempóváltozásról van szó. Ezért az `a_abs` időfüggvénye közvetlenül megmutatja, mikor volt dinamikailag erős a mozgás.

**Gyakori hiba:** Csak az egyik komponenssel, például `a_y`-nal dolgoznak, pedig a feladat az eredő gyorsulás nagyságát kéri.

**Feladatok:** Gy/4b, Gy/10b, Gy/12a, Gy/13c.


```python
t = np.arange(0.0, 80.0, 0.05)
sebesseg = 6.0 + 1.5*np.sin(2*np.pi*t/10.0)
iranyszog = 2*np.pi*t/18.0
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

mozgas = num_kinem(2)
mozgas.t = t
mozgas.r = np.column_stack((x, y))
mozgas.full_kinem_calc()

plt.figure(figsize=(8, 4))
plt.plot(mozgas.t, mozgas.a_abs)
plt.xlabel('idő (s)')
plt.ylabel('gyorsulás nagysága (m/s²)')
plt.grid()
plt.show()
```

### 3.5 Szélső pontok jelölése pályán

**Vizsgaszöveg:** Határozza meg azokat az időpontokat t = 3 s után, melyekre az autó a legközelebb illetve a legtávolabb került a kiindulás helyétől és jelölje be ezeket a test pályáján!

**Fizikai jelentés:** A kérdés a kiindulási ponttól mért pillanatnyi távolság szélsőértékeit keresi, nem a megtett út szélsőértékeit. A `dist` minden időpontra az origótól vett távolságot adja, ezért a `dist.argmin()` a legközelebbi, a `dist.argmax()` pedig a legtávolabbi pályapont indexét adja. Ugyanezekkel az indexekkel kell kiolvasni az időt és a helykoordinátákat, mert így ugyanarra a fizikai eseményre hivatkozunk.

**Gyakori hiba:** A `pathlength` minimumát vagy maximumát keresik, pedig a megtett út monoton nő, így nem alkalmas a kiindulástól való közelség/távolság vizsgálatára.

**Feladatok:** Gy/4e.

```python
phi = np.linspace(0.0, 2*np.pi, 800)
x = 15.0*np.cos(phi) + 6.0
y = 8.0*np.sin(phi) + 3.0

mozgas = num_kinem(2)
mozgas.t = np.linspace(0.0, 80.0, len(phi))
mozgas.r = np.column_stack((x, y))
mozgas.full_kinem_calc()

i_min = mozgas.dist.argmin()
i_max = mozgas.dist.argmax()

print(f'Legközelebb: t = {mozgas.t[i_min]:.2f} s, távolság = {mozgas.dist[i_min]:.2f} m')
print(f'Legtávolabb: t = {mozgas.t[i_max]:.2f} s, távolság = {mozgas.dist[i_max]:.2f} m')

plt.figure(figsize=(7, 7))
plt.plot(mozgas.r_x, mozgas.r_y, color='lightgray', label='pálya')
plt.scatter(0.0, 0.0, color='black', marker='+', s=100, label='kiindulási pont / origó')
plt.scatter(mozgas.r_x[i_min], mozgas.r_y[i_min], color='blue', marker='*', s=140, label='legközelebb')
plt.scatter(mozgas.r_x[i_max], mozgas.r_y[i_max], color='red', marker='*', s=140, label='legtávolabb')
plt.axis('equal')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.grid()
plt.show()
```

### 3.6 Paramétersöprés grafikonja

**Vizsgaszöveg:** Ábrázolja / határozza meg, milyen paraméter mellett maximális a vizsgált mennyiség!

**Fizikai jelentés:** A paraméter változtatása más kezdőfeltételt vagy modellt ad, ezért minden paraméterértékre külön kell kiszámítani a fizikai kimenetet. A grafikon megmutatja a kapcsolat alakját, az `argmax` pedig azt a paramétert választja ki, ahol a kimenet a legnagyobb.

**Gyakori hiba:** A ciklusban ugyanazt az objektumot használják újrainicializálás nélkül, így a korábbi futások állapota belekeveredik az új eredménybe.

**Feladatok:** Gy/6c, Gy/8c, Gy/8d, Gy/9e, Gy/11e, Gy/11f, Gy/12d, Gy/13d, Gy/14c, Gy/14d, F2-2026/f.


```python
def egyszeru_lotav(szog_fok):
    v0 = 16.0
    g = 1.625
    y0 = 1.8
    szog = np.radians(szog_fok)
    vx = v0*np.cos(szog)
    vy = v0*np.sin(szog)
    # Egyszerű, állandó holdi gravitációs közelítés a recept illusztrálására.
    repulesi_ido = (vy + np.sqrt(vy**2 + 2*g*y0)) / g
    return vx * repulesi_ido

szogek = np.linspace(5.0, 85.0, 81)
tavolsagok = np.array([egyszeru_lotav(szog) for szog in szogek])
i_max = tavolsagok.argmax()

plt.figure(figsize=(8, 4))
plt.plot(szogek, tavolsagok)
plt.scatter(szogek[i_max], tavolsagok[i_max], color='red', zorder=3)
plt.xlabel('paraméter: eldobási szög (fok)')
plt.ylabel('eredmény: dobástávolság (m)')
plt.grid()
plt.show()
```

### 3.7 Két feltétel egyidejű jelölése pályán

**Vizsgaszöveg:** Készítsen ábrát, melyen az autó pályája szürke, és rajta pirossal az erős kanyarodási részek, kékkel pedig a közel egyenes részek vannak bejelölve.

**Fizikai jelentés:** Két külön fizikai állapotot ugyanazon pályán akkor érdemes együtt jelölni, ha térben szeretnénk összehasonlítani, hol történnek. A kód két boolean maszkot készít ugyanarra a mozgásra, ezért a piros és kék pontok ugyanazon koordinátarendszerben értelmezhetők.

**Gyakori hiba:** Két külön ábrára rajzolják a feltételeket, így nehéz összehasonlítani, hogy a pálya mely részei tartoznak az egyes állapotokhoz.

**Feladatok:** Gy/7c.

```python
t = np.arange(0.0, 80.0, 0.05)
sebesseg = 12.0 + 3.0*np.sin(2*np.pi*t/20.0)
iranyszog = 0.5*np.sin(2*np.pi*t/12.0)
x = integ(t, sebesseg*np.cos(iranyszog))
y = integ(t, sebesseg*np.sin(iranyszog))

mozgas = num_kinem(2)
mozgas.t = t
mozgas.r = np.column_stack((x, y))
mozgas.full_kinem_calc()

piros = mozgas.a_cp_abs > 3.0
kek = mozgas.a_cp_abs < 0.5

plt.figure(figsize=(8, 4))
plt.plot(mozgas.r_x, mozgas.r_y, color='lightgray')
plt.scatter(mozgas.r_x[piros], mozgas.r_y[piros], color='red', s=6, label='erős kanyar')
plt.scatter(mozgas.r_x[kek], mozgas.r_y[kek], color='blue', s=6, label='közel egyenes')
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

### 3.8 Lokális csúcspontok táblázatos kiírása

**Vizsgaszöveg:** Készítsen táblázatot, mely a pálya lokálisan legmagasabb pontjaira adja meg sorra az időpontot, x és y koordinátát és a sebesség nagyságát!

**Fizikai jelentés:** A táblázat akkor jó forma, ha több hasonló eseményhez ugyanazokat a mennyiségeket kell megadni. A kód minden lokális csúcshoz ugyanazokat az oszlopokat írja ki, így az egymás utáni ívek könnyen összehasonlíthatók.

**Gyakori hiba:** Csak az ábrán jelölik a pontokat, de nem adják meg számszerűen az időt, helyet és sebességet.

**Feladatok:** Gy/9d.

```python
t = np.linspace(0.0, 10.0, 1000)
x = 2.0*t
y = np.maximum(0.0, 3.0*np.abs(np.sin(2*np.pi*t/4.0))*np.exp(-0.08*t))

mozgas = num_kinem(2)
mozgas.t = t
mozgas.r = np.column_stack((x, y))
mozgas.full_kinem_calc()

csucsok = np.where((mozgas.r_y[1:-1] > mozgas.r_y[:-2]) & (mozgas.r_y[1:-1] > mozgas.r_y[2:]))[0] + 1

print('sorszám\tidő (s)\tx (m)\ty (m)\tsebesség (m/s)')
for sorszam, i in enumerate(csucsok[:5], start=1):
    print(f'{sorszam}\t{mozgas.t[i]:.2f}\t{mozgas.r_x[i]:.2f}\t{mozgas.r_y[i]:.2f}\t{mozgas.v_abs[i]:.2f}')
```

### 3.9 Hárompaneles hely-sebesség-gyorsulás időgrafikon

**Vizsgaszöveg:** Ábrázolja egy pingpong-labda hely-, sebesség- és gyorsulás-idő függvényét a talajra érésig.

**Fizikai jelentés:** Egy mozgást gyakran három egymásból származó mennyiség ír le: hely, sebesség és gyorsulás. A közös időtengely azért hasznos, mert látszik, hogyan kapcsolódik a hely változása a sebességhez és a gyorsuláshoz.

**Gyakori hiba:** Három külön, eltérő skálájú ábrán rajzolják ki őket, így nehéz összevetni az időpontokat.

**Feladatok:** Gy/12a.

```python
t = np.linspace(0.0, 3.0, 400)
y = 25.0 - 0.5*9.81*t**2
y = np.maximum(y, 0.0)

mozgas = num_kinem(2)
mozgas.t = t
mozgas.r = np.column_stack((np.zeros_like(t), y))
mozgas.full_kinem_calc()

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 8), sharex=True)
ax1.plot(mozgas.t, mozgas.r_y)
ax1.set_ylabel('hely, y (m)')
ax1.grid()
ax2.plot(mozgas.t, mozgas.v_abs)
ax2.set_ylabel('sebesség (m/s)')
ax2.grid()
ax3.plot(mozgas.t, mozgas.a_abs)
ax3.set_ylabel('gyorsulás (m/s²)')
ax3.set_xlabel('idő (s)')
ax3.grid()
plt.show()
```

### 3.11 Lokális felső holtpontok jelölése pályán

**Vizsgaszöveg:** Készítsen ábrát, melyen a labda pályája szürkével, a felső holtpontok pedig kékkel vannak jelölve!

**Fizikai jelentés:** A felső holtpontok az egyes pattogási ívek lokális csúcsai, ahol a függőleges sebesség előjele pozitívról negatívra vált. A kód ezért először ezeket az indexeket keresi meg, majd ugyanazokat a pályapontokat színezi kékre. Így az ábra nem új számítást végez, hanem a már fizikailag azonosított eseményeket teszi láthatóvá.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba, hogy a globális legmagasabb pontot jelölik meg többször, nem pedig az egymást követő ívek külön csúcsait.

**Feladatok:** F2-2026/e.


```python
# a korábban megtalált felső holtpontokat
# ugyanazon a pályarajzon kék pontokkal jelöljük.
t = np.linspace(0.0, 16.0, 1601)
x = 1.8*t
y = 0.15 + 2.0*np.exp(-0.13*t)*np.maximum(np.sin(2.8*t), 0.0)

labda = num_kinem(2)
labda.t = t
labda.r = np.column_stack([x, y])
labda.full_kinem_calc()

# Magasság szerinti holtpont: v_y=0; függőleges irányban 0 a pill. seb.
ind_holtpont = arg_eq(labda.v[:, 1], 0)

# Csak valódi csúcsok: v_y pozitívból negatívba vált ÉS a magasság > 0.3 m
# Felső holtpont: olyan holtpont, ahol a_y<0, azaz lefelé mutat a gyorsulás.
# Magyarázat: a talajon pattanva felfelé gyorsul a test, azaz az alsó holtpontokban a_y>0
ind_felsoholtpont = ind_holtpont[
    (labda.a[ind_holtpont, 1] < 0) &
    (labda.r_y[ind_holtpont] > 0.3)  # kiszűri a talaj menti hamis holtpontokat
]

plt.figure(figsize=(9, 4))
plt.plot(labda.r_x, labda.r_y, color='0.65', label='pálya')
plt.scatter(labda.r_x[ind_felsoholtpont], labda.r_y[ind_felsoholtpont], color='blue', zorder=3,
            label='felső holtpontok')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.grid()
plt.legend()
plt.show()
```

### 3.12 Változó tömegű mozgási energia időgrafikonja

**Vizsgaszöveg:** Ábrázolja a rakéta mozgási energiáját az idő függvényében!

**Fizikai jelentés:** Változó tömegű testnél a mozgási energia időfüggése a sebesség és a tömeg együttes változásából jön. A grafikon azért válaszolja meg a kérdést, mert minden időponthoz a hozzá tartozó tömeget és sebességnagyságot használja. Így a hajtóanyag fogyása nem marad ki az energiából.

**Minimális kódrecept:** A következő kódcella közvetlenül erre a kérdésre ad minimális, önállóan futtatható megoldási mintát.

**Gyakori hiba:** Gyakori hiba, hogy `kinetic_energy_constant_m`-et használnak rakétára, pedig itt a tömeg nem állandó.

**Feladatok:** Gy/13b.


```python
t = np.linspace(0.0, 6.0, 601)
v = 4.0 + 28.0*(1.0 - np.exp(-t/1.8))
x = integ(t, v)
y = 1.0 + 0.6*t

raketa = num_kinem(2)
raketa.t = t
raketa.r = np.column_stack([x, y])
raketa.full_kinem_calc()

tomeg = np.maximum(0.15, 0.50 - 0.08*t)
E = raketa.kinetic_energy(tomeg)

plt.figure(figsize=(8, 4))
plt.plot(raketa.t, E)
plt.xlabel('idő (s)')
plt.ylabel('mozgási energia (J)')
plt.grid()
plt.show()
```
