```python
import quantities as pq 
```

```python
m=pq.UncertainQuantity(100, pq.gram, 0.8)
rho=pq.UncertainQuantity(8940, pq.kg/pq.m**3, 5)
```

Papíron kiszámolható:

$$m=\varrho\cdot V = \varrho\cdot \frac{4\pi}{3} r^3$$

Innen átrendezve:

$$r=\sqrt[3]{\frac{3 m}{\varrho\cdot4\pi}}$$

Ezért:

```python
r=(3*m/(rho*4*pq.pi))**(1/3)
```

```python
# A kérdés az átmérő volt:
d=2*r.simplified

print(f"A gyártandó gömbök átmérője: {d.magnitude:.4} m, a megengedett szórás: {d.uncertainty:.4}.")
```
