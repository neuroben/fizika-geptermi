```python
import numpy as np    
import matplotlib.pyplot as plt
%matplotlib inline    
 
from fizinfo import *
```

```python
def motor_r(t):
    phi=2*np.pi/12*t
    R=10+1*np.sin(2*np.pi/2.5*t)
    x=R*np.cos(phi)
    y=R*np.sin(phi)
    
    return np.array([x, y], np.float64)
```

```python
motor=num_kinem(2)   # 2D kinematikai számítások
motor.set_time_range(0.0, 100.0, 0.01)
```

```python
motor.set_r_fun(motor_r)
motor.calc_r_to_v()
motor.calc_v_to_a()
```

```python
# a)
```

```python
motor.plot_rcomp()
```

```python
# b)
```

```python
motor.v_abs=vect_abs(motor.v)
```

```python
vmax=motor.v_abs.max()
vmin=motor.v_abs.min()

print(f"A motor minimális sebessége: {vmin:.2f} m/s, maximális sebessége: {vmax:.2f} m/s.")
```

```python
# c)
```

```python
gyors = motor.v_abs>0.95*vmax 
lassu = motor.v_abs<1.05*vmin

fig=plt.figure(figsize=(8,8))  
ax1=plt.subplot(111)

ax1.plot(motor.r[:,0], motor.r[:,1], color="grey")  # a pályagörbe
# a pályagörbe nagy a_cp-jű részeinél adatpontok:
ax1.scatter(motor.r[gyors,0], motor.r[gyors,1], color="red", s=4)  
ax1.scatter(motor.r[lassu,0], motor.r[lassu,1], color="green", s=4) 
ax1.set_aspect(aspect='equal')
ax1.grid()
```
