### Pár gondolat a NumPy-ről

NumPy: Numerical Python modul. Szokásos hívása: `import numpy as np`.
  * https://numpy.org/
  * https://en.wikipedia.org/wiki/NumPy

Nagyon sokat tudó modul, mely lehetőséget ad Pythonban (relatíve) gyors numerikus számításokra. 

A használt rutinokat mindig felsorolom majd, rövid leírással, de az alapgondolatok a következők:

A NumPy alap adat-struktúrája a tömb (array), amik csupa azonos típusú elemből állnak. Főbb sajátosságok:
   * a listákhoz hasonlóan indexelhetők
   * támogatják az úgynevezett "vektorizált" végrehajtást, azaz amikor operátor vagy függvény automatikusan végrehajtódik minden elemre; pl. ha `arr1` és `arr2` két, azonos méretű NumPy tömb, akkor `arr1-arr2` egy ugyanekkora tömb, elemenként kiszámolva a különbséget
   * sok művelet egy paranccsal elérhető, pl. `arr.mean()` a tömbelemek átlaga, `arr.std()` a szórás (standard deviation)
   * sok algoritmus, pl. regresszió, interpoláció, egyenlet-megoldás, be van programozva
   
A használat többnyire elég kézenfekvő. Érdemes megtanulni, mert sokkal tömörebb és gyorsabb a kód, mintha Python list-ekkel és ciklusokkal dolgoznánk. (De egy C++-ban írt és lefordított program sebességét nem éri el.)

Sok más projekt épít erre, pl. SciPy (Scientific Python), OpenCV (Open Computer Vision), Pandas, ...

Ez a notebook a NumPy azon featureibe ad bevezetést, melyeket a felhasználunk a "Fizika informatikusoknak" tárgyban.

```python
import numpy as np                # a NumPy szokásos importálása
import matplotlib.pyplot as plt   # egyszerű grafikon rajzolás
%matplotlib inline    
```

## Első próbálkozások

```python
# Adott elemekből álló NumPy tömb (array) létrehozása:
aa=np.array([0.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0], dtype=np.float64)
```

A NumPy tömbök csak azonos típusú elemekből állhatnak. Ez jelenleg az np.float64 típus, azaz lebegőpontos számok.

```python
# A tömb fő paraméterei:
print("A tömb elemei:", aa)
print("A tömb mérete:", aa.shape)
print("A tömbelemek típusa:", aa.dtype)
```

Megjegyzés: a "shape" lehet többelemű is, ha a tömb nem 1, hanem többdimenziós. Egydimenziós esetben a `len()` függvény is visszaadja a méretet.

```python
# Egyszerű matematikai műveletek elemenként végrehajtódnak:
bb=aa*3
print(bb)
```

Elgondolkodtató kérdés: ha `aa` lista lenne, mi lenne a 3-szorosa?

**Vektorizálás**: Egy művelet végrehajtása egy tömb (vektor) összes elemére.

Nem nekünk kell leprogramozni a ciklusokat. Sokkal gyorsabb és tömörebb megoldás.

```python
# egyszerű aritmetikai művelet vektorizálása
cc=0.03*aa**2-3
print(cc)
```

```python
# Mindenféle függvény vektorizált verziója is benne van a NumPy-ban:
dd=100*np.sin(np.radians(aa))
print(dd)
```

```python
# Egyszerű grafikon rajzolás:

fig=plt.figure(figsize=(10,4))  # méret
ax1=fig.add_subplot(111) # csak egy rész-grafikon
ax1.set_xlabel("a")      # x tengely felirat
ax1.set_ylabel("f(a)")   # y tengely felirat
ax1.grid()               # rajzoljunk rácsot

ax1.plot(aa, bb, color="red", label="3*a")   # aa függvényében bb
ax1.plot(aa, cc, color="green", label="0.03*a**2-3")   # aa függvényében cc
ax1.plot(aa, dd, color="blue", label="100*sin(a°)")   # aa függvényében dd

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

## Kicsit haladóbb használat

```python
# Egyenletes lépésközű elemek NumPy tömbben:
t=np.arange(0.0, 5.0, 0.1, dtype=np.float64)   # np.arange(kezdet, vég, lépésköz, adattípus)
print(t)
```

```python
# majdnem ugyanez az eredmény:
t2=np.linspace(0.0, 5.0, 51)     # np.linspace(kezdet, vég, elemszám)
print(t2)
```

Kérdés: miben különbözik `t` és `t2`?

```python
# Egyből egy kis fizika: egyenes vonalú egyenletesen gyorsuló mozgás:
# Kezdő hely: 20, kezdő sebesség: 16, gyorsulás: -9.81
x= 20.0 + 16.0*t - (9.81/2)*t**2
print(x)  
# x: egy tömb, ugyanannyi elemmel, mint t.
# x tömb tartalma: az x(t)=20+16*t-(9.81/2)*t**2 függvény kiértékelve a t tömb minden pontjában.
```

```python
# Nézzük meg grafikonon:
fig=plt.figure(figsize=(10,4)) 
ax1=fig.add_subplot(111) 
ax1.set_xlabel("t [s]")   
ax1.set_ylabel("x(t) [m]")   
ax1.grid()               

ax1.plot(t, x, color="red", label="x(t)")   # t függvényében x

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

## Egyszerű, de hasznos keresések

```python
# x értékek maximuma:
print(f"Legnagyobb x érték: {x.max():.3f} m.")
```

```python
# Melyik tömbelemnél vette fel a maximumot?
print(f"Legnagyobb x érték helye a tömbben: {x.argmax():d} m.")
```

```python
# Mikor vette fel a maximumot?
ind_max=x.argmax()
print(f"Legnagyobb x értéket ennél a t-nél vette fel: {t[ind_max]:.3f} s.")
```

```python
# Mikor volt legközelebb az origóhoz?

ind_orig=np.abs(x-0.0).argmin()
t_orig=t[ind_orig]
print(f"Az origóhoz ekkor volt legközelebb: {t_orig:.3f} s.")
```

```python
# csupa 0 kezdőértékű tömb, mely ugyanakkora és ugyanolyan típusú, mint t:
ff=np.zeros_like(t)
print(ff)
```

## Tartomány-kijelölések és egy kis fizika

A NumPy tömbök indexeivel ugyanúgy kijelölhetünk részhalmazokat, mint a Python listák és stringek esetében.

```python
# Újra definiáljuk t és x tömbjét, hogy ne zavarjanak a korábbi számítások.
t=np.arange(0.0, 5.0, 0.1, np.float64)
# legyen egy mozgásunk (ugyanaz, mint korábban)
x=20.0 + 16.0*t - (9.81/2)*t**2
print(t)
print(t.shape)
print(x)
print(x.shape)
```

```python
# első és utolsó elemek:
t_elso=t[0]
t_utolso=t[-1]
teljes_delta_t=t[-1]-t[0]
print(f"Teljes időintervallum hossz: {teljes_delta_t:.2f}")
```

```python
# hasonlóan az elmozdulás:
teljes_delta_x=x[-1]-x[0]
print(f"Teljes elmozdulás: {teljes_delta_x:.2f}")
```

```python
# Átlagsebesség a teljes intervallumra: (fizika általános iskolából: átlagsebesség=elmozdulás/idő; v=Delta x/Delta t)
teljes_v_atl=teljes_delta_x/teljes_delta_t
print(f"Átlagsebesség a teljes mozgás során:: {teljes_v_atl:.2f}")
```

```python
# trükkösebb: szomszédok különbsége

t_nemelso=t[1:]   # t[1], t[2], ..., azaz lemarad az 0-s sorszámú
print("nem első elemek:", t_nemelso)
print("nem első elemek száma:", t_nemelso.shape)
```

```python
t_nemutolso=t[:-1]   # t[0], t[1], ..., t[utolsó előtti] azaz lemarad az utolsó, azaz a -1-es sorszámú
print("nem utolsó elemek:", t_nemutolso)
print("nem utolsó elemek száma:", t_nemutolso.shape)
```

```python
delta_t=t_nemelso-t_nemutolso # t szomészédos elemeinek különbsége
print(delta_t)
```

```python
# hasonlóan a delta_x-ek, de már tömören:
delta_x=x[1:]-x[:-1]

print(delta_x)
```

```python
# mivel ez gyakran kell, ezért van rá egy rutin:
delta_x=np.diff(x)

print(delta_x)
```

```python
# átlagsebesség szakaszról szakaszra:
v_atl=delta_x/delta_t
print(v_atl)

# Vigyázat! Ez eggyel kevesebb elemű, mint x!
```

```python
# ezek legjobban az intervalumok közepén levő sebességeket közelítik
t_kozep=(t[1:]+t[:-1])/2
print(t_kozep)
```

```python
fig=plt.figure(figsize=(10,4)) 
ax1=fig.add_subplot(111) 
ax1.set_xlabel("t [s]")     
ax1.set_ylabel("v(t) [m/s]")     
ax1.grid()              

ax1.plot(t_kozep, v_atl, color="orange", label="v(t)")   # t függvényében v_atl

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

```python
# Figyelem! delta_t, delta_x, v_atl egyel rövidebb, mint t és x!
```

```python
# út = kis elmodulások nagyságának összege
teljes_ut=np.abs(delta_x).sum()
# tömb.sum(): tömbelemek összege
print(f"Teljes út: {teljes_ut:.3f}")
```

```python
# Ha az összegzés részeredményei is érdekesek: tömb.cumsum()
ut=np.abs(delta_x).cumsum()

print(ut)
```

```python
# Ebből kimarad a 0. elem, de addig az út nyilván 0 volt. Ezért ez a korrekt:

ut=np.append([0], np.abs(delta_x).cumsum())
print(ut)
```

```python
fig=plt.figure(figsize=(10,4)) 
ax1=fig.add_subplot(111) 
ax1.set_xlabel("t [s]")     
ax1.set_ylabel("ut(t) [m]")      
ax1.grid()               

ax1.plot(t, ut, color="red", label="x(t)")   # t függvényében ut

ax1.legend(loc="upper left") # jelmagyarázat bal oldalt fent
```

## Az új eljárások gyűjteménye

Tömb létrehozás:
   * `aa=np.array(lista, dtype=típus)`: adott lista elemeit a típus-ra alakítja és ebből hoz létre tömböt
   * `aa=np.zeros(N, dtype=típus)`: `N` méretű, csupa 0-ból álló tömb
   * `bb=np.zeros_like(aa)`:  ugyanakkora és ugyanolyan típusú tömb létrehozása, mint `aa`
   * `cc=np.arange(start, end, delta, típus)`: `start` és `end` közt `delta` egyenletes lépésekkel létrejövő tömb
   * `cc2=np.linspace(start, end, N, típus)`: `start` és `end` közt egyenletes lépésekkel létrejövő `N` elemű tömb

Matemetikai függvények:
   * Műveleti jelek: `+`, `-`, `*`, `/`, `**`, ...: NumPy tömbökre elemenként vonatkoznak
   * `np.radians()`: fokról radiánra vált
   * `np.sin()`, `np.cos()`, `np.abs()`, ...: szinusz, koszinusz, abszolút érték, ...
   * `vfun=np.vectorize(fun)`: előállítja a `fun` vektorizált változatát

Összegzés:
   * `aa.sum()`: a tömb összes elemének összege
   * `aa.diff()`: szomszédok különbségéből álló tömb
   * `aa.cumsum()`: egy tömb fokozatos felösszegzése (kumulatív szumma)

Keresés:
   * `aa.max()`, `aa.min()`: a tömb maximális/minimális elemének nagysága
   * `aa.argmax()`, `aa.argmin()`: a tömb maximális/minimális elemének indexe (hányadik helyen van)
